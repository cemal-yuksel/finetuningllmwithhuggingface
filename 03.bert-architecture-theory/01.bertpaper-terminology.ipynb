## ğŸ§  1.4. Dil Modelleri ve BaÄŸlam

<div style="background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); padding: 20px; border-radius: 10px; color: white; margin: 20px 0;">
    <h3>ğŸ’­ Kritik Soru</h3>
    <p style="font-size: 16px;"><strong>Bir bilgisayar programÄ± (dil modeli) bu baÄŸlamÄ± nasÄ±l yakalayacak?</strong></p>
    <p style="font-size: 15px; margin-top: 10px;">Ä°ÅŸte tÃ¼m modern NLP'nin (DoÄŸal Dil Ä°ÅŸleme) temeli bu sorunun cevabÄ±nda yatÄ±yor!</p>
</div>

````python
# ğŸ¯ Basit BaÄŸlam Yakalama SimÃ¼lasyonu

def basit_baglam_analizi(cumle, hedef_kelime):
    """
    Ã‡ok basit bir ÅŸekilde kelimenin etrafÄ±ndaki kelimelere bakarak
    baÄŸlam ipuÃ§larÄ± bulmaya Ã§alÄ±ÅŸan bir fonksiyon.
    
    Bu gerÃ§ek bir dil modeli deÄŸil, sadece fikri gÃ¶stermek iÃ§in!
    """
    kelimeler = cumle.lower().split()
    
    try:
        index = kelimeler.index(hedef_kelime.lower())
    except ValueError:
        return "Kelime bulunamadÄ±!"
    
    # Ã–nceki ve sonraki kelimeleri al (baÄŸlam penceresi)
    onceki = kelimeler[max(0, index-2):index]
    sonraki = kelimeler[index+1:min(len(kelimeler), index+3)]
    
    return {
        "hedef": hedef_kelime,
        "onceki_baglan": " ".join(onceki) if onceki else "(baÅŸlangÄ±Ã§)",
        "sonraki_baglam": " ".join(sonraki) if sonraki else "(son)"
    }

# Test edelim
test_cumleleri = [
    ("Havuzda yÃ¼z Ã§ok iyiydi", "yÃ¼z"),
    ("Onun yÃ¼zÃ¼ bugÃ¼n Ã§ok solgundu", "yÃ¼zÃ¼"),
    ("YÃ¼z Ã¶ÄŸrenci derse girdi", "yÃ¼z")
]

print("\nğŸ”¬ BAÄLAM PENCERESÄ° ANALÄ°ZÄ°")
print("="*70)
print("(Modeller kelimeyi anlamak iÃ§in saÄŸÄ±na-soluna bakar!)\n")

for cumle, hedef in test_cumleleri:
    sonuc = basit_baglam_analizi(cumle, hedef)
    print(f"CÃ¼mle: \"{cumle}\"")
    print(f"  â¬…ï¸  Ã–nceki BaÄŸlam : [{sonuc['onceki_baglan']}]")
    print(f"  ğŸ¯  Hedef Kelime  : [{sonuc['hedef']}]")
    print(f"  â¡ï¸  Sonraki BaÄŸlam: [{sonuc['sonraki_baglam']}]")
    print("-"*70)
````

## ğŸ“Š 1.5. GÃ¶rselleÅŸtirme: BaÄŸlam Penceresi

AÅŸaÄŸÄ±daki gÃ¶rselleÅŸtirme, bir dil modelinin bir kelimeyi anlamak iÃ§in "baÄŸlam penceresine" nasÄ±l baktÄ±ÄŸÄ±nÄ± gÃ¶steriyor:

````python
# ğŸ“Š ASCII SanatÄ± ile BaÄŸlam Penceresi GÃ¶sterimi

def baglam_gorsellestir(cumle, hedef_kelime):
    """
    BaÄŸlam penceresini gÃ¶rsel olarak gÃ¶sterir
    """
    kelimeler = cumle.split()
    
    # Hedef kelimeyi bul
    hedef_lower = hedef_kelime.lower()
    hedef_index = -1
    
    for i, kelime in enumerate(kelimeler):
        if kelime.lower() == hedef_lower:
            hedef_index = i
            break
    
    if hedef_index == -1:
        print("Kelime bulunamadÄ±!")
        return
    
    print("\n" + "="*70)
    print(f"CÃ¼mle: {cumle}")
    print("="*70 + "\n")
    
    for i, kelime in enumerate(kelimeler):
        if i < hedef_index:
            print(f"  {kelime:>15} â¬…ï¸  [Ã–NCEKÄ° BAÄLAM]")
        elif i == hedef_index:
            print(f"\n  {kelime:>15} ğŸ¯  [HEDEF KELÄ°ME] â­\n")
        else:
            print(f"  {kelime:>15} â¡ï¸  [SONRAKÄ° BAÄLAM]")
    
    print("\n" + "="*70)

# Ã–rnekleri gÃ¶rselleÅŸtirelim
print("\nğŸ¨ BAÄLAM PENCERESÄ° GÃ–RSELLEÅTÄ°RMESÄ°")

baglam_gorsellestir("Havuzda yÃ¼z Ã§ok iyiydi", "yÃ¼z")
print("\nğŸ’¡ YORUM: 'Havuzda' kelimesi bize 'yÃ¼z' kelimesinin aktivite olduÄŸunu sÃ¶ylÃ¼yor!")

print("\n" + "*"*70 + "\n")

baglam_gorsellestir("Onun yÃ¼zÃ¼ bugÃ¼n Ã§ok solgundu", "yÃ¼zÃ¼")
print("\nğŸ’¡ YORUM: 'Onun' ve 'solgundu' kelimeleri bize 'yÃ¼zÃ¼' kelimesinin vÃ¼cut parÃ§asÄ± olduÄŸunu sÃ¶ylÃ¼yor!")
````

````python
# ğŸ¬ Tek YÃ¶nlÃ¼ BaÄŸlam SimÃ¼lasyonu

def tek_yonlu_baglam(cumle, hedef_index):
    """
    Tek yÃ¶nlÃ¼ modelin bakÄ±ÅŸ aÃ§Ä±sÄ±nÄ± simÃ¼le eder.
    Sadece hedef kelimeden Ã–NCEKÄ° kelimeleri gÃ¶rebilir!
    """
    kelimeler = cumle.split()
    
    print(f"\n{'='*70}")
    print(f"ğŸ¬ TEK YÃ–NLÃœ MODEL BAKIÅ AÃ‡ISI")
    print(f"{'='*70}\n")
    print(f"ğŸ“ Tam CÃ¼mle: {cumle}")
    print(f"ğŸ¯ Hedef Pozisyon: {hedef_index + 1}. kelime\n")
    print(f"{'â”€'*70}\n")
    
    for i, kelime in enumerate(kelimeler):
        if i < hedef_index:
            print(f"  âœ… {kelime:>15}  â† GÃ–REBÄ°LÄ°YOR (geÃ§miÅŸ)")
        elif i == hedef_index:
            print(f"\n  ğŸ¯ {kelime:>15}  â† ÅU AN BURADA\n")
        else:
            print(f"  âŒ {kelime:>15}  â† GÃ–REMÄ°YOR (gelecek - karartÄ±lmÄ±ÅŸ!)")
    
    print(f"\n{'='*70}")
    
    # Sadece Ã¶nceki kelimeleri dÃ¶ndÃ¼r
    onceki = kelimeler[:hedef_index]
    return {
        "gorulen_baglan": " ".join(onceki) if onceki else "(hiÃ§bir ÅŸey)",
        "hedef": kelimeler[hedef_index],
        "gorulemeyen": " ".join(kelimeler[hedef_index+1:]) if hedef_index+1 < len(kelimeler) else "(hiÃ§bir ÅŸey)"
    }

# Test edelim
ornek_cumle = "Onlar sokaÄŸÄ±n saÄŸ tarafÄ±ndaydÄ±"
sonuc = tek_yonlu_baglam(ornek_cumle, 2)  # "saÄŸ" kelimesi 3. sÄ±rada (index=2)

print(f"\nğŸ’¡ ANALÄ°Z:")
print(f"   Model sadece '{sonuc['gorulen_baglan']}' kÄ±smÄ±nÄ± gÃ¶rdÃ¼")
print(f"   '{sonuc['hedef']}' kelimesinin anlamÄ±nÄ± bu bilgiyle tahmin etmeli")
print(f"   Ama '{sonuc['gorulemeyen']}' kÄ±smÄ±nÄ± bilemiyor!")
print(f"\n   SonuÃ§: 'saÄŸ' kelimesinin YÃ–N mÃ¼, SAÄLIK mÄ±, SÄ°YASÄ° SAÄ mÄ±")
print(f"          olduÄŸunu tam anlayamÄ±yor! ğŸ˜•")
````

Film metaforuna geri dÃ¶nelim:
- ArtÄ±k ekranÄ±n tamamÄ±nÄ± gÃ¶rÃ¼yorsunuz! ğŸ“º

## ğŸ“Š 2.4. Diagram: Ã‡ift YÃ¶nlÃ¼ BaÄŸlam (Bidirectional)

```mermaid
flowchart LR
    A[Onlar] <--> C[saÄŸ]
    B[sokaÄŸÄ±n] <--> C

    E -.-> B

```

**AÃ§Ä±klama:**
- ğŸŸ¢ **Sol yeÅŸil kutular**: Ã–nceki kelimeler (soldan gelen baÄŸlam)
- ğŸŸ¡ **SarÄ± kutu**: Hedef kelime (anlamÄ±nÄ± bulmaya Ã§alÄ±ÅŸtÄ±ÄŸÄ±mÄ±z)
- ğŸ”µ **Mavi kutu**: Sonraki kelimeler (saÄŸdan gelen baÄŸlam)
- â†”ï¸ **Ã‡ift yÃ¶nlÃ¼ oklar**: Model HER Ä°KÄ° YÃ–NDEN de bilgi alÄ±yor!

<div style="background-color: #cfe2ff; border-left: 5px solid #0d6efd; padding: 15px; margin: 20px 0;">
    <h3>ğŸ¯ Kritik Fark</h3>
    <p style="font-size: 15px; margin: 0;">
        <strong>Tek YÃ¶nlÃ¼:</strong> Sadece â¬…ï¸ bu yÃ¶nden bakar<br>
        <strong>Ã‡ift YÃ¶nlÃ¼:</strong> Hem â¬…ï¸ hem â¡ï¸ her iki yÃ¶nden bakar!
    </p>
</div>

````python
# ğŸ”„ Ã‡ift YÃ¶nlÃ¼ BaÄŸlam SimÃ¼lasyonu (BERT TarzÄ±)

def cift_yonlu_baglam(cumle, hedef_index):
    """
    Ã‡ift yÃ¶nlÃ¼ modelin bakÄ±ÅŸ aÃ§Ä±sÄ±nÄ± simÃ¼le eder.
    Hem Ã¶nceki hem sonraki kelimeleri gÃ¶rebilir!
    """
    kelimeler = cumle.split()
    
    print(f"\n{'='*70}")
    print(f"ğŸ”„ Ã‡Ä°FT YÃ–NLÃœ MODEL BAKIÅ AÃ‡ISI (BERT)")
    print(f"{'='*70}\n")
    print(f"ğŸ“ Tam CÃ¼mle: {cumle}")
    print(f"ğŸ¯ Hedef Pozisyon: {hedef_index + 1}. kelime\n")
    print(f"{'â”€'*70}\n")
    
    for i, kelime in enumerate(kelimeler):
        if i < hedef_index:
            print(f"  âœ… {kelime:>15}  â¬…ï¸  GÃ–REBÄ°LÄ°YOR (sol baÄŸlam)")
        elif i == hedef_index:
            print(f"\n  ğŸ¯ {kelime:>15}  â­ ÅU AN BURADA (HER Ä°KÄ° YÃ–NDEN BAKABÄ°LÄ°YOR!)\n")
        else:
            print(f"  âœ… {kelime:>15}  â¡ï¸  GÃ–REBÄ°LÄ°YOR (saÄŸ baÄŸlam)")
    
    print(f"\n{'='*70}")
    
    # Hem Ã¶nceki hem sonraki kelimeleri dÃ¶ndÃ¼r
    onceki = kelimeler[:hedef_index]
    sonraki = kelimeler[hedef_index+1:]
    
    return {
        "sol_baglam": " ".join(onceki) if onceki else "(baÅŸlangÄ±Ã§)",
        "hedef": kelimeler[hedef_index],
        "sag_baglam": " ".join(sonraki) if sonraki else "(son)"
    }

# Test edelim - AynÄ± cÃ¼mleyi kullan
ornek_cumle = "Onlar sokaÄŸÄ±n saÄŸ tarafÄ±ndaydÄ±"
sonuc = cift_yonlu_baglam(ornek_cumle, 2)  # "saÄŸ" kelimesi

print(f"\nğŸ’¡ ANALÄ°Z:")
print(f"   â¬…ï¸  Sol BaÄŸlam : '{sonuc['sol_baglam']}'")
print(f"   ğŸ¯  Hedef Kelime: '{sonuc['hedef']}'")
print(f"   â¡ï¸  SaÄŸ BaÄŸlam : '{sonuc['sag_baglam']}'")
print(f"\n   âœ¨ SONUÃ‡:")
print(f"      â€¢ 'sokaÄŸÄ±n' kelimesi â†’ mekan belirtiyor")
print(f"      â€¢ 'tarafÄ±ndaydÄ±' kelimesi â†’ yÃ¶n belirtiyor")
print(f"      â€¢ Ä°ki taraftan gelen bilgi birleÅŸince:")
print(f"        'saÄŸ' kelimesi = YÃ–N anlamÄ±nda! âœ…")
print(f"\n   ğŸ‰ Ã‡ift yÃ¶nlÃ¼ baÄŸlam sayesinde anlam kesinleÅŸti!")
````

````python
# ğŸ® Ä°nteraktif Test: Kendi CÃ¼mlelerinizi Deneyin!

def baglan_analiz_et(cumle, hedef_kelime_index):
    """
    Tek yÃ¶nlÃ¼ ve Ã§ift yÃ¶nlÃ¼ modellerin aynÄ± cÃ¼mleyi nasÄ±l 
    farklÄ± ÅŸekilde iÅŸlediÄŸini gÃ¶sterir.
    """
    kelimeler = cumle.split()
    
    if hedef_kelime_index >= len(kelimeler) or hedef_kelime_index < 0:
        print("âŒ Hata: GeÃ§ersiz kelime indeksi!")
        return
    
    hedef = kelimeler[hedef_kelime_index]
    onceki = kelimeler[:hedef_kelime_index]
    sonraki = kelimeler[hedef_kelime_index+1:]
    
    print("\n" + "="*80)
    print("ğŸ”¬ BAÄLAM ANALÄ°Z SÄ°STEMÄ°")
    print("="*80)
    print(f"\nğŸ“ CÃ¼mle: {cumle}")
    print(f"ğŸ¯ Analiz edilen kelime: '{hedef}' (pozisyon: {hedef_kelime_index+1})")
    print("\n" + "â”€"*80)
    
    # Tek YÃ¶nlÃ¼
    print("\nğŸ¬ TEK YÃ–NLÃœ MODEL GÃ–RÃœÅÃœ:")
    print("   " + "â”Œ" + "â”€"*70 + "â”")
    if onceki:
        print(f"   â”‚ GÃ¶rÃ¼len: {' '.join(onceki):<63} â”‚")
    else:
        print(f"   â”‚ GÃ¶rÃ¼len: (hiÃ§bir ÅŸey){' '*50} â”‚")
    print(f"   â”‚ Hedef  : {hedef:<63} â”‚")
    print(f"   â”‚ Durum  : âš ï¸  SÄ±nÄ±rlÄ± bilgi - tahmin zor!{' '*32} â”‚")
    print("   â””" + "â”€"*70 + "â”˜")
    
    # Ã‡ift YÃ¶nlÃ¼
    print("\nğŸ”„ Ã‡Ä°FT YÃ–NLÃœ MODEL GÃ–RÃœÅÃœ (BERT):")
    print("   " + "â”Œ" + "â”€"*70 + "â”")
    sol = ' '.join(onceki) if onceki else "(baÅŸlangÄ±Ã§)"
    sag = ' '.join(sonraki) if sonraki else "(son)"
    print(f"   â”‚ Sol    : {sol:<63} â”‚")
    print(f"   â”‚ Hedef  : {hedef:<63} â”‚")
    print(f"   â”‚ SaÄŸ    : {sag:<63} â”‚")
    print(f"   â”‚ Durum  : âœ… Tam baÄŸlam - anlam net!{' '*34} â”‚")
    print("   â””" + "â”€"*70 + "â”˜")
    
    print("\n" + "="*80)
    
    # Analiz
    print("\nğŸ’¡ SONUÃ‡ ANALÄ°ZÄ°:")
    print(f"   â€¢ Tek yÃ¶nlÃ¼ model sadece {len(onceki)} Ã¶nceki kelimeyi gÃ¶rdÃ¼")
    print(f"   â€¢ Ã‡ift yÃ¶nlÃ¼ model {len(onceki)} Ã¶nceki + {len(sonraki)} sonraki = {len(onceki)+len(sonraki)} baÄŸlam kelimesi gÃ¶rdÃ¼")
    print(f"   â€¢ Ã‡ift yÃ¶nlÃ¼ {int((1 - len(onceki)/(len(onceki)+len(sonraki)+1))*100) if len(onceki)+len(sonraki) > 0 else 0}% daha fazla bilgiye sahip!")
    print("\n" + "="*80 + "\n")

# Test Ã¶rnekleri
print("ğŸ¯ Ã–RNEK 1: BankacÄ±lÄ±k Senaryosu")
baglan_analiz_et("MÃ¼ÅŸteri hesap iÅŸlemi baÅŸarÄ±yla tamamlandÄ±", 2)

print("\n\nğŸ¯ Ã–RNEK 2: Ã‡ok AnlamlÄ± Kelime")
baglan_analiz_et("Projenin son aÅŸamasÄ± Ã§ok kritik", 1)

print("\nğŸ“ NOT: YukarÄ±daki fonksiyonu kendi cÃ¼mlelerinizle test edebilirsiniz!")
print("Ã–rnek: baglan_analiz_et('Kendi cÃ¼mleniz buraya', hedef_kelime_sÄ±rasÄ±)")
print("(Kelime sÄ±rasÄ± 0'dan baÅŸlar!)")
````

````python
# ğŸ­ MLM SimÃ¼lasyonu: TÃ¼rkÃ§e Ã–rnekler

import random

def mlm_demo(cumle, mask_ratio=0.15):
    """
    MLM (Masked Language Modeling) sÃ¼recini simÃ¼le eder.
    
    Args:
        cumle: Maskelenecek cÃ¼mle
        mask_ratio: KaÃ§ kelimeyi maskeleyeceÄŸiz (varsayÄ±lan %15)
    """
    kelimeler = cumle.split()
    n_mask = max(1, int(len(kelimeler) * mask_ratio))
    
    # Rastgele kelime seÃ§ (gerÃ§ek BERT'te daha karmaÅŸÄ±k bir strateji var)
    mask_indices = random.sample(range(len(kelimeler)), n_mask)
    
    # Orijinal kelimeleri sakla
    orijinal_kelimeler = {}
    
    print("="*90)
    print("ğŸ­ MLM (MASKED LANGUAGE MODELING) SÄ°MÃœLASYONU")
    print("="*90)
    print(f"\nğŸ“ Orijinal CÃ¼mle:")
    print(f"   {cumle}")
    print(f"\nğŸ¯ Maskelenecek Kelime SayÄ±sÄ±: {n_mask} (yaklaÅŸÄ±k %{int(mask_ratio*100)})")
    
    # Maskeleme
    masked_kelimeler = kelimeler.copy()
    for idx in mask_indices:
        orijinal_kelimeler[idx] = kelimeler[idx]
        masked_kelimeler[idx] = "[MASK]"
    
    masked_cumle = " ".join(masked_kelimeler)
    print(f"\nğŸ­ MaskelenmiÅŸ CÃ¼mle:")
    print(f"   {masked_cumle}")
    
    # BaÄŸlam analizi
    print(f"\nğŸ” Her [MASK] Ä°Ã§in BaÄŸlam Analizi:")
    print("â”€"*90)
    
    for idx in sorted(mask_indices):
        orijinal = orijinal_kelimeler[idx]
        sol_baglam = " ".join(kelimeler[max(0, idx-2):idx]) if idx > 0 else "(baÅŸlangÄ±Ã§)"
        sag_baglam = " ".join(kelimeler[idx+1:min(len(kelimeler), idx+3)]) if idx < len(kelimeler)-1 else "(son)"
        
        print(f"\nğŸ¯ Pozisyon {idx+1}:")
        print(f"   â¬…ï¸  Sol BaÄŸlam  : '{sol_baglam}'")
        print(f"   ğŸ­  [MASK]      : ???")
        print(f"   â¡ï¸  SaÄŸ BaÄŸlam : '{sag_baglam}'")
        print(f"   ğŸ’¡ BERT Tahmini: '{orijinal}' âœ…")
        print(f"   ğŸ”— MantÄ±k      : Ã‡ift yÃ¶nlÃ¼ baÄŸlamdan tahmin edildi")
    
    print("\n" + "="*90)
    print("âœ… SonuÃ§: BERT bu ÅŸekilde milyonlarca cÃ¼mle iÅŸleyerek dili Ã¶ÄŸrenir!")
    print("="*90 + "\n")
    
    return masked_cumle, orijinal_kelimeler

# TÃ¼rkÃ§e YBS Ã¶rnekleri
ornekler = [
    "BandÄ±rma Onyedi EylÃ¼l Ãœniversitesi'nde YBS okuyan Ã¶ÄŸrenciler veri analizi dersini seviyor",
    "Kredi kartÄ± iÅŸlemi ÅŸÃ¼pheli bulunduÄŸu iÃ§in hesap incelemeye alÄ±ndÄ±",
    "MÃ¼ÅŸteri ÅŸikayet oranÄ± son iki haftada arttÄ± ve Ã§aÄŸrÄ± merkezi planÄ± deÄŸiÅŸti"
]

print("\nğŸ“ YBS Ã–ÄRENCÄ°LERÄ° Ä°Ã‡Ä°N MLM Ã–RNEKLERÄ°\n")

for i, ornek in enumerate(ornekler, 1):
    print(f"\n{'#'*90}")
    print(f"Ã–RNEK {i}:")
    print(f"{'#'*90}\n")
    mlm_demo(ornek)
    
    if i < len(ornekler):
        print("\n" + "."*90 + "\n")
````

## ğŸ¯ 4.3. BERT'in Maskeleme Stratejisi (15% KuralÄ±)

BERT, eÄŸitim sÄ±rasÄ±nda kelimelerin **%15'ini** maskeler. Ama bu %15'in iÃ§inde Ã¼Ã§ farklÄ± strateji kullanÄ±r:

<div style="background-color: #e7f3ff; border: 2px solid #2196F3; border-radius: 10px; padding: 20px; margin: 20px 0;">
    <h3 style="color: #1976D2;">ğŸ“‹ Maskeleme Stratejisi DetayÄ±</h3>
    
**%15'lik kelimelerin:**
- ğŸ­ **%80'i â†’ [MASK] token'Ä± ile deÄŸiÅŸtirilir**
  - Ã–rnek: "veri analizi" â†’ "veri [MASK]"
  
- ğŸ”„ **%10'u â†’ Rastgele baÅŸka bir kelime ile deÄŸiÅŸtirilir**
  - Ã–rnek: "veri analizi" â†’ "veri balÄ±k" (saÃ§ma ama kasÄ±tlÄ±!)
  - AmaÃ§: Modelin baÄŸlama gÃ¼venmesini saÄŸlamak
  
- âœ… **%10'u â†’ OlduÄŸu gibi bÄ±rakÄ±lÄ±r (deÄŸiÅŸmez)**
  - Ã–rnek: "veri analizi" â†’ "veri analizi"
  - AmaÃ§: Modelin gerÃ§ek kelime daÄŸÄ±lÄ±mÄ±nÄ± Ã¶ÄŸrenmesi
</div>

````python
# ğŸ¯ BERT'in GerÃ§ek Maskeleme Stratejisi

import random

def bert_maskeleme_stratejisi(cumle, mask_ratio=0.15):
    """
    BERT'in gerÃ§ek maskeleme stratejisini simÃ¼le eder:
    - %80 [MASK] ile deÄŸiÅŸtir
    - %10 rastgele kelime ile deÄŸiÅŸtir  
    - %10 olduÄŸu gibi bÄ±rak
    """
    kelimeler = cumle.split()
    n_mask = max(1, int(len(kelimeler) * mask_ratio))
    
    # Rastgele pozisyonlar seÃ§
    mask_positions = random.sample(range(len(kelimeler)), n_mask)
    
    masked_kelimeler = kelimeler.copy()
    degisiklikler = []
    
    # Rastgele kelimeler havuzu (gerÃ§ekÃ§i olmak iÃ§in)
    rastgele_kelimeler = ["masa", "kedi", "araba", "deniz", "bulut", "kalem", "kitap", "pencere"]
    
    for pos in mask_positions:
        orijinal = kelimeler[pos]
        rand_val = random.random()
        
        if rand_val < 0.8:  # %80
            masked_kelimeler[pos] = "[MASK]"
            strateji = "ğŸ­ [MASK]"
        elif rand_val < 0.9:  # %10
            rastgele_kelime = random.choice(rastgele_kelimeler)
            masked_kelimeler[pos] = rastgele_kelime
            strateji = f"ğŸ”„ Rastgele: '{rastgele_kelime}'"
        else:  # %10
            # OlduÄŸu gibi bÄ±rak
            strateji = "âœ… DeÄŸiÅŸmedi"
        
        degisiklikler.append({
            "pozisyon": pos + 1,
            "orijinal": orijinal,
            "yeni": masked_kelimeler[pos],
            "strateji": strateji
        })
    
    print("="*100)
    print("ğŸ¯ BERT'Ä°N GERÃ‡EK MASKELEME STRATEJÄ°SÄ°")
    print("="*100)
    print(f"\nğŸ“ Orijinal CÃ¼mle:\n   {cumle}")
    print(f"\nğŸ­ MaskelenmiÅŸ CÃ¼mle:\n   {' '.join(masked_kelimeler)}")
    
    print(f"\nğŸ“Š Maskeleme DetaylarÄ±:")
    print("â”€"*100)
    
    mask_count = sum(1 for d in degisiklikler if "[MASK]" in d["strateji"])
    random_count = sum(1 for d in degisiklikler if "Rastgele" in d["strateji"])
    keep_count = sum(1 for d in degisiklikler if "DeÄŸiÅŸmedi" in d["strateji"])
    
    for degisiklik in degisiklikler:
        print(f"\nğŸ”¹ Pozisyon {degisiklik['pozisyon']}:")
        print(f"   Orijinal: '{degisiklik['orijinal']}'")
        print(f"   Yeni     : '{degisiklik['yeni']}'")
        print(f"   Strateji : {degisiklik['strateji']}")
    
    print("\n" + "â”€"*100)
    print(f"\nğŸ“ˆ Strateji DaÄŸÄ±lÄ±mÄ±:")
    print(f"   ğŸ­ [MASK] kullanÄ±ldÄ±        : {mask_count} kez ({mask_count/len(degisiklikler)*100:.0f}%)")
    print(f"   ğŸ”„ Rastgele kelime kondu    : {random_count} kez ({random_count/len(degisiklikler)*100:.0f}%)")
    print(f"   âœ… DeÄŸiÅŸtirilmeden bÄ±rakÄ±ldÄ±: {keep_count} kez ({keep_count/len(degisiklikler)*100:.0f}%)")
    
    print("\nğŸ’¡ Neden Bu Strateji?")
    print("   â€¢ [MASK] â†’ Model boÅŸluk doldurma Ã¶ÄŸrenir")
    print("   â€¢ Rastgele â†’ Model baÄŸlama gÃ¼venmeyi Ã¶ÄŸrenir (gÃ¼rÃ¼ltÃ¼ye dayanÄ±klÄ±lÄ±k)")
    print("   â€¢ DeÄŸiÅŸmez â†’ Model gerÃ§ek kelime daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶rÃ¼r")
    
    print("\n" + "="*100 + "\n")

# Test edelim
ybs_cumlesi = "MÃ¼ÅŸteri ÅŸikayet oranÄ± son iki haftada arttÄ± ve Ã§aÄŸrÄ± merkezi vardiya planÄ± yeniden dÃ¼zenlendi"
bert_maskeleme_stratejisi(ybs_cumlesi)

print("\nğŸ”„ Not: Her Ã§alÄ±ÅŸtÄ±rÄ±ÅŸta rastgele farklÄ± sonuÃ§lar Ã¼retir!")
print("ğŸ“ Bu BERT'in eÄŸitim sÃ¼recindeki gerÃ§ek davranÄ±ÅŸÄ±nÄ± simÃ¼le eder.")
````

## ğŸ“Š 5.2. NSP SÃ¼reci: AdÄ±m AdÄ±m

````python
# ğŸ”— NSP SimÃ¼lasyonu: YBS Ã–rnekleri

import random

def nsp_demo(cumle_a, cumle_b, gercek_devam=True):
    """
    NSP (Next Sentence Prediction) sÃ¼recini simÃ¼le eder.
    
    Args:
        cumle_a: Ä°lk cÃ¼mle
        cumle_b: Ä°kinci cÃ¼mle
        gercek_devam: B, A'nÄ±n gerÃ§ek devamÄ± mÄ±?
    """
    print("="*100)
    print("ğŸ”— NSP (NEXT SENTENCE PREDICTION) SÄ°MÃœLASYONU")
    print("="*100)
    
    # BERT formatÄ±: [CLS] CÃ¼mle A [SEP] CÃ¼mle B [SEP]
    bert_input = f"[CLS] {cumle_a} [SEP] {cumle_b} [SEP]"
    
    print(f"\nğŸ“ CÃ¼mle A:")
    print(f"   {cumle_a}")
    print(f"\nğŸ“ CÃ¼mle B:")
    print(f"   {cumle_b}")
    
    print(f"\nğŸ”„ BERT Girdi FormatÄ±:")
    print(f"   {bert_input}")
    
    print(f"\nğŸ¯ GerÃ§ek Durum:")
    if gercek_devam:
        print(f"   âœ… IsNext (B, A'nÄ±n doÄŸal devamÄ±)")
    else:
        print(f"   âŒ NotNext (B, A ile alakasÄ±z)")
    
    # Basit bir tahmin simÃ¼lasyonu (gerÃ§ekte BERT aÄŸÄ±rlÄ±klarÄ±nÄ± kullanÄ±r)
    tahmin = gercek_devam  # SimÃ¼lasyonda doÄŸru tahmin
    confidence = random.uniform(0.85, 0.99) if tahmin == gercek_devam else random.uniform(0.40, 0.60)
    
    print(f"\nğŸ¤– BERT Tahmini:")
    if tahmin:
        print(f"   âœ… IsNext (gÃ¼ven: {confidence:.2%})")
    else:
        print(f"   âŒ NotNext (gÃ¼ven: {confidence:.2%})")
    
    print(f"\nğŸ’¡ Analiz:")
    if tahmin == gercek_devam:
        print(f"   âœ… Tahmin DOÄRU!")
        print(f"   ğŸ“‰ Loss: dÃ¼ÅŸÃ¼k (model Ã¶ÄŸreniyor)")
    else:
        print(f"   âŒ Tahmin YANLIÅ!")
        print(f"   ğŸ“ˆ Loss: yÃ¼ksek (model dÃ¼zeltilecek)")
    
    print("\n" + "="*100 + "\n")

# YBS Ã–rnekleri - DoÄŸal Devam (IsNext)
print("\n" + "ğŸ“ YBS Ã–RNEKLERÄ° - DOÄAL DEVAM (IsNext)".center(100))
print("="*100 + "\n")

ornekler_devam = [
    {
        "a": "MÃ¼ÅŸteri ÅŸikayet oranÄ± son iki haftada arttÄ±.",
        "b": "Bu nedenle Ã§aÄŸrÄ± merkezi vardiya planÄ± yeniden dÃ¼zenlendi.",
        "devam": True
    },
    {
        "a": "Kredi kartÄ± iÅŸlemi ÅŸÃ¼pheli bulundu.",
        "b": "Hesap gÃ¼venlik nedeniyle geÃ§ici olarak donduruldu.",
        "devam": True
    },
    {
        "a": "YÄ±llÄ±k satÄ±ÅŸ hedefi %95 oranÄ±nda gerÃ§ekleÅŸti.",
        "b": "YÃ¶netim ekibe prim Ã¶dÃ¼lÃ¼ vermeye karar verdi.",
        "devam": True
    }
]

for i, ornek in enumerate(ornekler_devam, 1):
    print(f"{'#'*100}")
    print(f"Ã–RNEK {i}:")
    print(f"{'#'*100}\n")
    nsp_demo(ornek["a"], ornek["b"], ornek["devam"])

# YBS Ã–rnekleri - Devam DeÄŸil (NotNext)
print("\n" + "ğŸ“ YBS Ã–RNEKLERÄ° - DEVAM DEÄÄ°L (NotNext)".center(100))
print("="*100 + "\n")

ornekler_devam_degil = [
    {
        "a": "MÃ¼ÅŸteri ÅŸikayet oranÄ± son iki haftada arttÄ±.",
        "b": "BugÃ¼n hava Ã§ok gÃ¼neÅŸli ve sÄ±cak.",
        "devam": False
    },
    {
        "a": "Kredi kartÄ± iÅŸlemi ÅŸÃ¼pheli bulundu.",
        "b": "TavÅŸanlar havuÃ§ yemeyi sever.",
        "devam": False
    }
]

for i, ornek in enumerate(ornekler_devam_degil, 1):
    print(f"{'#'*100}")
    print(f"Ã–RNEK {len(ornekler_devam) + i}:")
    print(f"{'#'*100}\n")
    nsp_demo(ornek["a"], ornek["b"], ornek["devam"])
````

````python
# ğŸ² OlasÄ±lÄ±ksal Dil Modelleme SimÃ¼lasyonu

import random

def olasiliksal_tahmin(cumle, maske_index):
    """
    BERT'Ä±n olasÄ±lÄ±ksal tahmin sÃ¼recini simÃ¼le eder.
    
    Args:
        cumle: MaskelenmiÅŸ cÃ¼mle (liste)
        maske_index: Maskenin indeksi
    """
    print("="*100)
    print("ğŸ² OLASILIKSAL DÄ°L MODELLEME SÄ°MÃœLASYONU")
    print("="*100)
    
    cumle_str = " ".join(cumle)
    print(f"\nğŸ“ GiriÅŸ CÃ¼mlesi:")
    print(f"   {cumle_str}")
    
    # YBS baÄŸlamÄ±na gÃ¶re aday kelimeler ve olasÄ±lÄ±klar
    if "mÃ¼ÅŸteri" in cumle_str.lower() or "hizmet" in cumle_str.lower():
        adaylar = {
            "memnuniyeti": 0.45,
            "ÅŸikayetleri": 0.25,
            "talepleri": 0.15,
            "geri": 0.08,
            "deneyimi": 0.05,
            "diÄŸerleri": 0.02
        }
    elif "satÄ±ÅŸ" in cumle_str.lower() or "gelir" in cumle_str.lower():
        adaylar = {
            "hedefi": 0.40,
            "performansÄ±": 0.30,
            "rakamlarÄ±": 0.15,
            "verisi": 0.10,
            "grafiÄŸi": 0.04,
            "diÄŸerleri": 0.01
        }
    elif "gÃ¼venlik" in cumle_str.lower() or "ÅŸifre" in cumle_str.lower():
        adaylar = {
            "protokolÃ¼": 0.35,
            "aÃ§Ä±ÄŸÄ±": 0.30,
            "seviyesi": 0.20,
            "sistemi": 0.10,
            "testi": 0.04,
            "diÄŸerleri": 0.01
        }
    else:
        adaylar = {
            "Ã§ok": 0.45,
            "hÄ±zla": 0.30,
            "son": 0.15,
            "Ã¶nemli": 0.08,
            "bÃ¼yÃ¼k": 0.02
        }
    
    print(f"\nğŸ§  BERT Ã‡ift YÃ¶nlÃ¼ Analiz YapÄ±yor...")
    print(f"   â¬…ï¸  Sol baÄŸlam: {' '.join(cumle[:maske_index])}")
    print(f"   ğŸ¯ [MASK]")
    print(f"   â¡ï¸  SaÄŸ baÄŸlam: {' '.join(cumle[maske_index+1:])}")
    
    print(f"\nğŸ“Š OlasÄ±lÄ±k DaÄŸÄ±lÄ±mÄ± (Softmax SonuÃ§larÄ±):")
    print(f"{'â”€'*100}")
    
    # OlasÄ±lÄ±klarÄ± sÄ±rala ve gÃ¶rselleÅŸtir
    sirali_adaylar = sorted(adaylar.items(), key=lambda x: x[1], reverse=True)
    
    for i, (kelime, olasilik) in enumerate(sirali_adaylar, 1):
        # Renk gradyanÄ±
        if i == 1:
            renk = "ğŸ¥‡"
            bar_color = "ğŸŸ©" * int(olasilik * 50)
        elif i == 2:
            renk = "ğŸ¥ˆ"
            bar_color = "ğŸŸ¦" * int(olasilik * 50)
        elif i == 3:
            renk = "ğŸ¥‰"
            bar_color = "ğŸŸ¨" * int(olasilik * 50)
        else:
            renk = f"{i}."
            bar_color = "â¬œ" * int(olasilik * 50)
        
        # YÃ¼zde gÃ¶sterimi
        yuzde = olasilik * 100
        
        print(f"   {renk} {kelime:15s} | {yuzde:5.1f}% | {bar_color}")
    
    print(f"{'â”€'*100}")
    print(f"   {'TOPLAM:':17s} | {sum(adaylar.values())*100:5.1f}% (Softmax garantisi)")
    
    # En yÃ¼ksek olasÄ±lÄ±klÄ± kelimeyi seÃ§
    en_iyi_kelime = sirali_adaylar[0][0]
    en_iyi_olasilik = sirali_adaylar[0][1]
    
    print(f"\nğŸ¯ BERT'Ä±n Tahmini:")
    print(f"   âœ… SeÃ§ilen kelime: '{en_iyi_kelime}' (gÃ¼ven: {en_iyi_olasilik*100:.1f}%)")
    
    # SonuÃ§ cÃ¼mlesi
    sonuc_cumle = cumle.copy()
    sonuc_cumle[maske_index] = en_iyi_kelime
    print(f"\nğŸ’¬ TamamlanmÄ±ÅŸ CÃ¼mle:")
    print(f"   {' '.join(sonuc_cumle)}")
    
    print("\n" + "="*100 + "\n")

# YBS Ã–rnekleri
print("\n" + "ğŸ“ YBS Ã–RNEKLERÄ°: OLASILIKSAL TAHMÄ°N".center(100))
print("="*100 + "\n")

ybs_ornekleri = [
    {
        "cumle": ["MÃ¼ÅŸteri", "hizmet", "kalitesi", "[MASK]", "deÄŸerlendirildi"],
        "maske_idx": 3
    },
    {
        "cumle": ["SatÄ±ÅŸ", "ekibi", "[MASK]", "aÅŸmayÄ±", "baÅŸardÄ±"],
        "maske_idx": 2
    },
    {
        "cumle": ["Siber", "gÃ¼venlik", "[MASK]", "tespit", "edildi"],
        "maske_idx": 2
    }
]

for i, ornek in enumerate(ybs_ornekleri, 1):
    print(f"{'#'*100}")
    print(f"Ã–RNEK {i}:")
    print(f"{'#'*100}\n")
    olasiliksal_tahmin(ornek["cumle"], ornek["maske_idx"])
````
