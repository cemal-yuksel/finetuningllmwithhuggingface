````python
# ğŸ“Š SÄ±nÄ±f daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶rselleÅŸtirme

# Her duygu iÃ§in Ã¶rnek sayÄ±sÄ±nÄ± hesaplÄ±yoruz
label_counts = {}
for example in train_data:
    label = example['label']
    label_counts[label] = label_counts.get(label, 0) + 1

# Duygu isimleri ve sayÄ±larÄ±
emotions = [emotion_labels[i] for i in sorted(label_counts.keys())]
counts = [label_counts[i] for i in sorted(label_counts.keys())]

# Pastel renkler (Her duygu iÃ§in Ã¶zel renk!)
colors_palette = ['#74b9ff', '#55efc4', '#fab1a0', '#fd79a8', '#a29bfe', '#ffeaa7']

# ğŸ“Š Grafik oluÅŸturma
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Sol grafik: Bar Chart
ax1 = axes[0]
bars = ax1.bar(range(len(emotions)), counts, color=colors_palette, 
               edgecolor='white', linewidth=2, alpha=0.8)

# Bar chart'Ä± sÃ¼slÃ¼yoruz
ax1.set_xlabel('Duygu Kategorileri', fontsize=13, fontweight='bold')
ax1.set_ylabel('Ã–rnek SayÄ±sÄ±', fontsize=13, fontweight='bold')
ax1.set_title('EÄŸitim Setindeki Duygu DaÄŸÄ±lÄ±mÄ±', fontsize=15, fontweight='bold', pad=20)
ax1.set_xticks(range(len(emotions)))
ax1.set_xticklabels(emotions, rotation=45, ha='right', fontsize=10)
ax1.grid(axis='y', alpha=0.3, linestyle='--')

# Bar'larÄ±n Ã¼zerine sayÄ±larÄ± yazÄ±yoruz
for i, (bar, count) in enumerate(zip(bars, counts)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{count:,}',
             ha='center', va='bottom', fontweight='bold', fontsize=11)

# SaÄŸ grafik: Pie Chart (Pasta GrafiÄŸi)
ax2 = axes[1]

# Pasta grafiÄŸi Ã¶zellikleri
wedges, texts, autotexts = ax2.pie(
    counts, 
    labels=emotions,
    colors=colors_palette,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05] * len(emotions),  # TÃ¼m parÃ§alarÄ± biraz ayÄ±rÄ±yoruz
    shadow=True,
    textprops={'fontsize': 10, 'weight': 'bold'}
)

# Pasta grafiÄŸini gÃ¼zelleÅŸtiriyoruz
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(11)

ax2.set_title('Duygu DaÄŸÄ±lÄ±mÄ± YÃ¼zdeleri', fontsize=15, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

# Ä°statistikleri yazdÄ±rÄ±yoruz
print("\nğŸ“ˆ DUYGU DAÄILIMI Ä°STATÄ°STÄ°KLERÄ°:")
print("="*70)
for emotion, count in zip(emotions, counts):
    percentage = (count / sum(counts)) * 100
    bar_length = int(percentage / 2)  # Bar uzunluÄŸu
    bar = 'â–ˆ' * bar_length + 'â–‘' * (50 - bar_length)
    print(f"{emotion:<25} {bar} {count:>5,} ({percentage:>5.1f}%)")
print("="*70)
print(f"{'TOPLAM':<25} {'â–ˆ' * 50} {sum(counts):>5,} (100.0%)")
print("="*70)

# Dengesizlik kontrolÃ¼
max_count = max(counts)
min_count = min(counts)
imbalance_ratio = max_count / min_count

print(f"\nâš–ï¸ Veri Dengesi Analizi:")
print(f"   En Fazla : {max(zip(counts, emotions))[1]} ({max_count:,} Ã¶rnek)")
print(f"   En Az    : {min(zip(counts, emotions))[1]} ({min_count:,} Ã¶rnek)")
print(f"   Oran     : {imbalance_ratio:.2f}x")

if imbalance_ratio < 2:
    print("   âœ… Veri seti oldukÃ§a dengeli!")
elif imbalance_ratio < 5:
    print("   âš ï¸ Orta seviye dengesizlik var, ama yÃ¶netilebilir.")
else:
    print("   ğŸš¨ Ciddi dengesizlik var, Ã¶zel teknikler gerekebilir.")
````

````python
# ğŸ“ Metin uzunluklarÄ±nÄ± analiz ediyoruz

# TÃ¼m metinlerin uzunluklarÄ±nÄ± hesaplÄ±yoruz (karakter bazÄ±nda)
text_lengths = [len(example['text']) for example in train_data]

# Ä°statistikler
mean_length = np.mean(text_lengths)
median_length = np.median(text_lengths)
max_length = np.max(text_lengths)
min_length = np.min(text_lengths)
std_length = np.std(text_lengths)

# GÃ¶rselleÅŸtirme
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Sol grafik: Histogram
ax1 = axes[0]
n, bins, patches = ax1.hist(text_lengths, bins=50, color='#74b9ff', 
                             edgecolor='white', alpha=0.8, linewidth=1.5)

# Renk gradyanÄ± ekliyoruz
cm = plt.cm.get_cmap('RdYlGn_r')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))

ax1.axvline(mean_length, color='#e74c3c', linestyle='--', linewidth=2.5, 
            label=f'Ortalama: {mean_length:.1f}')
ax1.axvline(median_length, color='#2ecc71', linestyle='--', linewidth=2.5, 
            label=f'Medyan: {median_length:.1f}')

ax1.set_xlabel('Metin UzunluÄŸu (Karakter)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Frekans', fontsize=12, fontweight='bold')
ax1.set_title('ğŸ“Š Metin Uzunluk DaÄŸÄ±lÄ±mÄ±', fontsize=14, fontweight='bold', pad=15)
ax1.legend(fontsize=11)
ax1.grid(axis='y', alpha=0.3, linestyle='--')

# SaÄŸ grafik: Box Plot
ax2 = axes[1]
box = ax2.boxplot(text_lengths, vert=True, patch_artist=True,
                   boxprops=dict(facecolor='#a29bfe', alpha=0.7),
                   medianprops=dict(color='#e74c3c', linewidth=2.5),
                   whiskerprops=dict(color='#2d3436', linewidth=1.5),
                   capprops=dict(color='#2d3436', linewidth=1.5),
                   flierprops=dict(marker='o', markerfacecolor='#fd79a8', 
                                  markersize=6, alpha=0.5))

ax2.set_ylabel('Metin UzunluÄŸu (Karakter)', fontsize=12, fontweight='bold')
ax2.set_title('ğŸ“¦ Box Plot - Uzunluk DaÄŸÄ±lÄ±mÄ±', fontsize=14, fontweight='bold', pad=15)
ax2.set_xticklabels(['TÃ¼m Metinler'])
ax2.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

# DetaylÄ± istatistikler
print("\nğŸ“Š METÄ°N UZUNLUK Ä°STATÄ°STÄ°KLERÄ°:")
print("="*60)
print(f"ğŸ“ Ortalama Uzunluk    : {mean_length:.2f} karakter")
print(f"ğŸ“ Medyan Uzunluk      : {median_length:.2f} karakter")
print(f"ğŸ“ˆ Maksimum Uzunluk    : {max_length} karakter")
print(f"ğŸ“‰ Minimum Uzunluk     : {min_length} karakter")
print(f"ğŸ“Š Standart Sapma     : {std_length:.2f}")
print("="*60)

# YÃ¼zdelik dilimler (Percentiles)
percentiles = [25, 50, 75, 90, 95, 99]
print("\nğŸ“ˆ YÃœZDELÄ°K DÄ°LÄ°MLER:")
print("="*60)
for p in percentiles:
    value = np.percentile(text_lengths, p)
    print(f"   {p}. YÃ¼zdelik: {value:.0f} karakter")
print("="*60)

# BERT iÃ§in Ã¶neri
print("\nğŸ’¡ BERT Ä°Ã‡Ä°N Ã–NERÄ°:")
print("="*60)
print(f"   BERT maksimum token limiti: 512")
print(f"   Ã–nerilen max_length: {int(np.percentile(text_lengths, 95) * 1.3)}")
print(f"   (Karakterlerin ~%95'ini kapsayacak ÅŸekilde)")
print("="*60)

# En uzun ve en kÄ±sa metinleri gÃ¶sterelim
print("\nğŸ“ Ã–RNEK METÄ°NLER:")
print("="*60)

# En kÄ±sa metin
shortest_idx = text_lengths.index(min_length)
print(f"\nğŸ”¹ En KÄ±sa Metin ({min_length} karakter):")
print(f"   '{train_data[shortest_idx]['text']}'")
print(f"   Duygu: {emotion_labels[train_data[shortest_idx]['label']]}")

# En uzun metin
longest_idx = text_lengths.index(max_length)
print(f"\nğŸ”¹ En Uzun Metin ({max_length} karakter):")
print(f"   '{train_data[longest_idx]['text']}'")
print(f"   Duygu: {emotion_labels[train_data[longest_idx]['label']]}")
print("="*60)
````

## ğŸ”¤ **AdÄ±m 4.4: N-gram Analizi**

    <p style="color: #2d3436; font-size: 1.05em; line-height: 1.8;">
        <b>N-gram</b>, ardÄ±ÅŸÄ±k olarak gelen <b>N</b> kelime kombinasyonudur.<br>
        ğŸ“Œ <b>Bigram (2-gram):</b> 2 kelimelik kombinasyonlar ("very good", "not happy")<br>
        ğŸ“Œ <b>Trigram (3-gram):</b> 3 kelimelik kombinasyonlar ("i love you", "makes me happy")

### ğŸ“Š **N-gram Analizi Ä°ÅŸ DÃ¼nyasÄ±nda Neden Ã–nemli?**

```mermaid
graph LR
    A[ğŸ”¤ N-gram Analizi] --> B[ğŸ“Š Trend Tespiti]
    A --> C[ğŸ’¬ YaygÄ±n Ä°fadeler]
    A --> D[ğŸ¯ MÃ¼ÅŸteri Dili]
    
    B --> E[ğŸ’¼ Strateji]

    E --> F[âœ¨ Ä°ÅŸ KararlarÄ±]
```

    <p style="font-size: 1.05em; color: #1976d2;">
        ğŸ’¡ <b>Ã–rnek:</b> "not good" bigram'Ä±, "not" ve "good" kelimelerinden daha fazla anlam taÅŸÄ±r!
    </p>
</div>

````python
# ğŸ”¤ Bigram ve Trigram analizi

def get_ngrams(texts, n=2, top_k=15):
    """
    N-gram'larÄ± Ã§Ä±karan ve sÄ±klÄ±ÄŸÄ±nÄ± hesaplayan fonksiyon
    
    Args:
        texts: Metin listesi
        n: N-gram boyutu (2=bigram, 3=trigram)
        top_k: Ä°lk kaÃ§ n-gram dÃ¶ndÃ¼rÃ¼lecek
    
    Returns:
        En sÄ±k kullanÄ±lan n-gram'lar
    """
    ngrams_list = []
    
    for text in texts:
        # Metni kÃ¼Ã§Ã¼k harfe Ã§evir ve kelimelere ayÄ±r
        words = re.findall(r'\b[a-z]+\b', text.lower())
        # Stopword'leri filtrele
        words = [w for w in words if w not in stop_words and len(w) > 2]
        
        # N-gram'larÄ± oluÅŸtur
        for i in range(len(words) - n + 1):
            ngram = ' '.join(words[i:i+n])
            ngrams_list.append(ngram)
    
    # SÄ±klÄ±k hesaplama
    ngram_freq = Counter(ngrams_list).most_common(top_k)
    return ngram_freq

print("ğŸ”¤ N-gram analizi yapÄ±lÄ±yor...\n")
print("="*70)

# TÃ¼m metinleri liste olarak hazÄ±rlama
all_texts_list = [example['text'] for example in train_data]

# BIGRAM ANALÄ°ZÄ° (2'li kelime kombinasyonlarÄ±)
print("\nğŸ“Š EN SIK KULLANILAN 15 BÄ°GRAM (2'li kelime):")
print("="*70)

bigrams = get_ngrams(all_texts_list, n=2, top_k=15)

for i, (bigram, count) in enumerate(bigrams, 1):
    bar_length = int(count / bigrams[0][1] * 40)
    bar = 'â–ˆ' * bar_length + 'â–‘' * (40 - bar_length)
    print(f"{i:2}. {bigram:<20} {bar} {count:>5,}")

# TRIGRAM ANALÄ°ZÄ° (3'lÃ¼ kelime kombinasyonlarÄ±)
print("\n\nğŸ“Š EN SIK KULLANILAN 15 TRÄ°GRAM (3'lÃ¼ kelime):")
print("="*70)

trigrams = get_ngrams(all_texts_list, n=3, top_k=15)

for i, (trigram, count) in enumerate(trigrams, 1):
    bar_length = int(count / trigrams[0][1] * 40)
    bar = 'â–ˆ' * bar_length + 'â–‘' * (40 - bar_length)
    print(f"{i:2}. {trigram:<30} {bar} {count:>4,}")

print("\n" + "="*70)
print("âœ… N-gram analizi tamamlandÄ±!")
print("="*70)
````

````python
# ğŸ“ Optimal max_length deÄŸerini bulma

print("ğŸ“ Token uzunluÄŸu analizi yapÄ±lÄ±yor...\n")
print("="*60)

# TÃ¼m metinleri tokenize edip uzunluklarÄ±nÄ± bulma
token_lengths = []

print("ğŸ”„ Veri seti tokenize ediliyor (bu biraz zaman alabilir)...")

for example in tqdm(train_data, desc="Tokenizing"):
    # Metni tokenize et
    tokens = tokenizer.tokenize(example['text'])
    token_lengths.append(len(tokens))

# Ä°statistikler
mean_tokens = np.mean(token_lengths)
median_tokens = np.median(token_lengths)
max_tokens = np.max(token_lengths)
min_tokens = np.min(token_lengths)

print("\n" + "="*60)
print("ğŸ“Š TOKEN UZUNLUK Ä°STATÄ°STÄ°KLERÄ°:")
print("="*60)
print(f"ğŸ“ Ortalama: {mean_tokens:.1f} token")
print(f"ğŸ“ Medyan: {median_tokens:.1f} token")
print(f"ğŸ“ˆ Maksimum: {max_tokens} token")
print(f"ğŸ“‰ Minimum: {min_tokens} token")
print("="*60)

# YÃ¼zdelik dilimler
print("\nğŸ“Š YÃœZDELÄ°K DÄ°LÄ°MLER:")
print("="*60)
percentiles = [50, 75, 90, 95, 99]
for p in percentiles:
    value = np.percentile(token_lengths, p)
    print(f"   {p:2}. yÃ¼zdelik: {value:>6.1f} token")
print("="*60)

# GÃ¶rselleÅŸtirme
fig, ax = plt.subplots(1, 1, figsize=(14, 6))

# Histogram
n, bins, patches = ax.hist(token_lengths, bins=50, color='#74b9ff', 
                            edgecolor='white', alpha=0.8, linewidth=1.5)

# Renk gradyanÄ±
cm = plt.cm.get_cmap('RdYlGn_r')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))

# Ä°statistik Ã§izgileri
ax.axvline(mean_tokens, color='#e74c3c', linestyle='--', linewidth=2.5, 
           label=f'Ortalama: {mean_tokens:.1f}')
ax.axvline(median_tokens, color='#2ecc71', linestyle='--', linewidth=2.5, 
           label=f'Medyan: {median_tokens:.1f}')
ax.axvline(np.percentile(token_lengths, 95), color='#f39c12', linestyle='--', 
           linewidth=2.5, label=f'95. yÃ¼zdelik: {np.percentile(token_lengths, 95):.1f}')

ax.set_xlabel('Token SayÄ±sÄ±', fontsize=12, fontweight='bold')
ax.set_ylabel('Frekans', fontsize=12, fontweight='bold')
ax.set_title('ğŸ“Š Token Uzunluk DaÄŸÄ±lÄ±mÄ±', fontsize=14, fontweight='bold', pad=15)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

# Ã–nerilen max_length
recommended_max_length = int(np.percentile(token_lengths, 95)) + 2  # +2 for [CLS] and [SEP]
print(f"\nğŸ’¡ Ã–NERÄ°LEN MAX_LENGTH: {recommended_max_length}")
print(f"   Bu deÄŸer metinlerin %95'ini kapsayacaktÄ±r.")
print("="*60)

# Max_length'i belirleme
if recommended_max_length > 512:
    max_length = 512
    print(f"\nâš ï¸ Ã–nerilen deÄŸer 512'den bÃ¼yÃ¼k, BERT limiti olan 512 kullanÄ±lacak.")
else:
    max_length = min(recommended_max_length, 128)  # HÄ±z iÃ§in 128 ile sÄ±nÄ±rlÄ±yoruz
    print(f"\nâœ… Max length olarak {max_length} kullanÄ±lacak.")

print("="*60)
````

## ğŸ”„ **AdÄ±m 5.3: Preprocessing Fonksiyonu OluÅŸturma**

<div style="background: linear-gradient(135deg, #55efc4 0%, #81ecec 100%); padding: 20px; border-radius: 10px; margin: 20px 0;">
    <h3 style="color: #00695c;">âš™ï¸ Preprocessing Pipeline</h3>
    <p style="color: #00695c; font-size: 1.05em;">
        TÃ¼m veri setini tek bir fonksiyonla iÅŸleyeceÄŸiz!<br>
        Hugging Face Datasets kÃ¼tÃ¼phanesi bunu Ã§ok hÄ±zlÄ± yapÄ±yor! ğŸš€
    </p>

### ğŸ“¦ **Fonksiyon Parametreleri**

```mermaid
graph TD
    A[ğŸ“ Preprocessing Fonksiyonu] --> B[ğŸ”¤ Tokenization]
    A --> C[ğŸ“ Max Length]
    A --> D[âœ‚ï¸ Truncation]
    A --> E[â• Padding]
    
    B --> F[Token ID'leri]
    C --> F
    D --> F
    E --> F
    
    F --> G[ğŸ¯ Attention Mask]
    G --> H[âœ… HazÄ±r Dataset!]
    
    style A fill:#ffeaa7,stroke:#fdcb6e,stroke-width:3px
    style B fill:#74b9ff,stroke:#0984e3,stroke-width:2px
    style G fill:#55efc4,stroke:#00b894,stroke-width:2px
    style H fill:#fab1a0,stroke:#e17055,stroke-width:3px
```

````python
# ğŸ”„ Preprocessing fonksiyonu tanÄ±mlama

def preprocess_function(examples):
    """
    Veri setini BERT iÃ§in hazÄ±rlayan fonksiyon
    
    Args:
        examples: Batch halinde gelen veri Ã¶rnekleri
        
    Returns:
        Tokenize edilmiÅŸ ve iÅŸlenmiÅŸ veri
    """
    # Tokenization
    # padding='max_length': TÃ¼m metinleri max_length'e kadar doldurur
    # truncation=True: max_length'den uzun metinleri keser
    # max_length: BelirlediÄŸimiz maksimum uzunluk
    # return_tensors=None: Åimdilik liste olarak dÃ¶ndÃ¼r (sonra tensor'a Ã§evrilecek)
    
    return tokenizer(
        examples['text'],
        padding='max_length',
        truncation=True,
        max_length=max_length,
        return_tensors=None
    )

print("âœ… Preprocessing fonksiyonu tanÄ±mlandÄ±!")
print("="*60)

# Fonksiyonu test edelim
print("\nğŸ§ª Fonksiyon Testi:")
print("="*60)

# Ä°lk 3 Ã¶rneÄŸi alalÄ±m
test_examples = {
    'text': [train_data[i]['text'] for i in range(3)],
    'label': [train_data[i]['label'] for i in range(3)]
}

print(f"ğŸ“ Test metinleri ({len(test_examples['text'])} adet):")
for i, text in enumerate(test_examples['text'], 1):
    print(f"   {i}. {text[:50]}...")

# Preprocessing
processed = preprocess_function(test_examples)

print(f"\nâœ… Ä°ÅŸlenmiÅŸ veri:")
print(f"   ğŸ”‘ Anahtarlar: {list(processed.keys())}")
print(f"   ğŸ“ input_ids boyutu: {len(processed['input_ids'][0])} token")
print(f"   ğŸ¯ attention_mask boyutu: {len(processed['attention_mask'][0])} token")

# Ä°lk Ã¶rneÄŸin detayÄ±
print(f"\nğŸ” Ä°lk Ã–rnek DetayÄ±:")
print(f"   ğŸ“ Orijinal: '{test_examples['text'][0]}'")
print(f"   ğŸ”¢ input_ids: {processed['input_ids'][0][:20]}... (ilk 20 token)")
print(f"   ğŸ¯ attention_mask: {processed['attention_mask'][0][:20]}... (ilk 20)")

print("\n" + "="*60)
print("âœ¨ Fonksiyon baÅŸarÄ±yla Ã§alÄ±ÅŸÄ±yor!")
print("="*60)
````

````python
# ğŸ“¦ Veri setine preprocessing uygulama

print("ğŸš€ TÃ¼m veri setine tokenization uygulanÄ±yor...")
print("="*70)
print("â³ Bu iÅŸlem biraz zaman alabilir, lÃ¼tfen bekleyin...")
print("="*70)
print()

# map() fonksiyonu ile tÃ¼m veri setine preprocessing uygula
tokenized_dataset = dataset.map(
    preprocess_function,
    batched=True,                    # Toplu iÅŸlem yap (hÄ±zlÄ±!)
    batch_size=1000,                 # Her seferde 1000 Ã¶rnek iÅŸle
    remove_columns=['text'],         # 'text' sÃ¼tununu kaldÄ±r (artÄ±k gerek yok)
    desc="ğŸ”„ Tokenizing"             # Ä°lerleme Ã§ubuÄŸu aÃ§Ä±klamasÄ±
)

print()
print("="*70)
print("âœ… TÃœM VERÄ° SETÄ° BAÅARIYLA TOKENÄ°ZE EDÄ°LDÄ°!")
print("="*70)
print()

# SonuÃ§larÄ± gÃ¶ster
print("ğŸ“Š Ä°ÅŸlenmiÅŸ Veri Seti Ä°statistikleri:")
print("="*70)

for split_name in tokenized_dataset.keys():
    split_data = tokenized_dataset[split_name]
    print(f"\nğŸ”¹ {split_name.upper()}")
    print(f"   ğŸ“ Ã–rnek sayÄ±sÄ±: {len(split_data):,}")
    print(f"   ğŸ”‘ SÃ¼tunlar: {split_data.column_names}")
    print(f"   ğŸ“ Ã–zellikler: {split_data.features}")

print()
print("="*70)
print("ğŸ¯ Veri seti model eÄŸitimi iÃ§in tamamen hazÄ±r!")
print("="*70)
````

### ğŸ¯ **Evaluation Metrics**

Model performansÄ±nÄ± Ã¶lÃ§mek iÃ§in **4 temel metrik** kullanacaÄŸÄ±z:

| **Metrik** | **FormÃ¼l** | **AÃ§Ä±klama** |
|------------|------------|--------------|
| **Accuracy** | DoÄŸru tahminler / Toplam | Genel baÅŸarÄ± oranÄ± |
| **Precision** | TP / (TP + FP) | Pozitif tahminlerin doÄŸruluÄŸu |
| **Recall** | TP / (TP + FN) | GerÃ§ek pozitifleri bulma oranÄ± |
| **F1-Score** | 2 Ã— (P Ã— R) / (P + R) | Precision ve Recall'un harmonik ortalamasÄ± |

```mermaid

    style G fill:#fab1a0,stroke:#e17055,stroke-width:2px
```

ğŸ”§ Metrik hesaplama fonksiyonunu yazalÄ±m!

````python
# ğŸ“Š Metrik fonksiyonlarÄ± tanÄ±mlama
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(eval_pred):
    """
    Evaluation sÄ±rasÄ±nda metrikleri hesaplayan fonksiyon
    
    Args:
        eval_pred: EvalPrediction nesnesi (predictions, label_ids iÃ§erir)
        
    Returns:
        Dictionary: accuracy, precision, recall, f1 metrikleri
    """
    # Predictions ve labels
    predictions, labels = eval_pred
    
    # Logits'ten tahminleri al (en yÃ¼ksek skora sahip sÄ±nÄ±f)
    predictions = np.argmax(predictions, axis=1)
    
    # Metrikleri hesapla
    # average='weighted': Her sÄ±nÄ±fÄ±n Ã¶rnek sayÄ±sÄ±na gÃ¶re aÄŸÄ±rlÄ±klandÄ±rÄ±lmÄ±ÅŸ ortalama
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, 
        predictions, 
        average='weighted',
        zero_division=0
    )
    
    # Accuracy hesapla
    accuracy = accuracy_score(labels, predictions)
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

print("âœ… Metrik fonksiyonu tanÄ±mlandÄ±!")
print("="*70)
print()

# Fonksiyonu test edelim
print("ğŸ§ª FONKSIYON TESTÄ°:")
print("-"*70)

# Dummy predictions ve labels oluÅŸtur
dummy_predictions = np.array([
    [2.1, 0.5, 3.2, 1.0, 0.8, 1.5],  # Predicted: 2
    [1.0, 4.5, 0.3, 1.2, 0.9, 0.5],  # Predicted: 1
    [0.8, 1.2, 0.5, 3.8, 1.0, 0.6],  # Predicted: 3
    [2.5, 0.9, 4.0, 1.1, 0.7, 1.3],  # Predicted: 2
])
dummy_labels = np.array([2, 1, 3, 2])

# EvalPrediction benzeri tuple oluÅŸtur
from collections import namedtuple
EvalPrediction = namedtuple('EvalPrediction', ['predictions', 'label_ids'])
dummy_eval_pred = EvalPrediction(dummy_predictions, dummy_labels)

# Metrikleri hesapla
test_metrics = compute_metrics(dummy_eval_pred)

print(f"ğŸ“Š Test Metrikleri:")
for metric_name, metric_value in test_metrics.items():
    print(f"   â€¢ {metric_name:12s}: {metric_value:.4f}")

print()
print("="*70)
print("âœ… Metrik fonksiyonu test edildi ve Ã§alÄ±ÅŸÄ±yor!")
print("="*70)
````

````python
# âš™ï¸ TrainingArguments ayarlama
from transformers import TrainingArguments

print("âš™ï¸ TRAINING ARGUMENTS AYARLANIYOR...")
print("="*70)
print()

# Training arguments tanÄ±mlama
training_args = TrainingArguments(
    # Temel ayarlar
    output_dir="./bert-emotion-classifier",          # Model checkpoint'leri buraya kaydedilecek
    overwrite_output_dir=True,                       # Var olan klasÃ¶rÃ¼n Ã¼zerine yaz
    
    # EÄŸitim hiperparametreleri
    num_train_epochs=3,                              # 3 epoch eÄŸitim
    per_device_train_batch_size=8,                   # Her GPU/CPU iÃ§in 8 Ã¶rnek
    per_device_eval_batch_size=16,                   # Evaluation iÃ§in 16 Ã¶rnek (daha hÄ±zlÄ±)
    
    # Optimizer ayarlarÄ±
    learning_rate=2e-5,                              # BERT iÃ§in optimal learning rate
    weight_decay=0.01,                               # L2 regularization (overfitting Ã¶nleme)
    warmup_steps=500,                                # Ä°lk 500 adÄ±mda learning rate'i yavaÅŸÃ§a artÄ±r
    
    # Loglama ve kaydetme
    logging_dir='./logs',                            # TensorBoard loglarÄ±
    logging_steps=100,                               # Her 100 adÄ±mda log yaz
    logging_strategy="steps",                        # Her N adÄ±mda log
    
    # Evaluation stratejisi
    eval_strategy="epoch",                           # Her epoch sonunda evaluate et
    save_strategy="epoch",                           # Her epoch sonunda model kaydet
    
    # Model seÃ§imi
    load_best_model_at_end=True,                     # EÄŸitim bitince en iyi modeli yÃ¼kle
    metric_for_best_model="f1",                      # F1-score'a gÃ¶re en iyi model
    greater_is_better=True,                          # YÃ¼ksek F1 daha iyi
    
    # Performans optimizasyonlarÄ±
    fp16=False,                                      # Mixed precision (GPU iÃ§in)
    dataloader_num_workers=0,                        # Veri yÃ¼kleme iÃ§in thread sayÄ±sÄ±
    
    # Seed
    seed=42,                                         # Reproducibility iÃ§in
    
    # DiÄŸer
    report_to="none",                                # Wandb, tensorboard vb. entegrasyonu kapalÄ±
    push_to_hub=False,                               # Hugging Face Hub'a yÃ¼kleme
)

print("âœ… TrainingArguments baÅŸarÄ±yla oluÅŸturuldu!")
print("="*70)
print()

# AyarlarÄ± gÃ¶rÃ¼ntÃ¼le
print("ğŸ“‹ TRAINING ARGUMENTS DETAYLARI:")
print("-"*70)
print(f"ğŸ“ Output Directory: {training_args.output_dir}")
print(f"ğŸ”¢ Epochs: {training_args.num_train_epochs}")
print(f"ğŸ“¦ Train Batch Size: {training_args.per_device_train_batch_size}")
print(f"ğŸ“¦ Eval Batch Size: {training_args.per_device_eval_batch_size}")
print(f"ğŸ“ˆ Learning Rate: {training_args.learning_rate}")
print(f"âš–ï¸  Weight Decay: {training_args.weight_decay}")
print(f"ğŸ”¥ Warmup Steps: {training_args.warmup_steps}")
print(f"ğŸ“Š Logging Steps: {training_args.logging_steps}")
print(f"âœ… Eval Strategy: {training_args.eval_strategy}")
print(f"ğŸ’¾ Save Strategy: {training_args.save_strategy}")
print(f"ğŸ† Best Model Metric: {training_args.metric_for_best_model}")
print(f"ğŸ² Random Seed: {training_args.seed}")
print()

# Toplam training steps hesaplama
train_dataset_size = len(tokenized_dataset['train'])
steps_per_epoch = train_dataset_size // training_args.per_device_train_batch_size
total_steps = steps_per_epoch * training_args.num_train_epochs

print("ğŸ“Š TRAINING Ä°STATÄ°STÄ°KLERÄ°:")
print("-"*70)
print(f"ğŸ”¢ Training Ã–rnekleri: {train_dataset_size:,}")
print(f"ğŸ”¢ Steps per Epoch: {steps_per_epoch:,}")
print(f"ğŸ”¢ Total Training Steps: {total_steps:,}")
print(f"ğŸ”¥ Warmup Steps: {training_args.warmup_steps} ({training_args.warmup_steps/total_steps*100:.1f}% of total)")
print()

print("="*70)
print("âœ… EÄŸitim konfigÃ¼rasyonu hazÄ±r!")
print("="*70)
````

- âœ… Hangi duygularÄ± doÄŸru tahmin ettiÄŸimiz

````python
# ğŸ­ Confusion Matrix oluÅŸturma
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("ğŸ­ CONFUSION MATRIX OLUÅTURULUYOR...")
print("="*70)
print()

# Test seti iÃ§in tahminleri al
predictions_output = trainer.predict(tokenized_dataset['test'])
predictions = np.argmax(predictions_output.predictions, axis=1)
true_labels = predictions_output.label_ids

print(f"ğŸ“Š Toplam Test Ã–rneÄŸi: {len(true_labels)}")
print(f"ğŸ¯ Tahmin SayÄ±sÄ±: {len(predictions)}")
print()

# Confusion matrix hesapla
cm = confusion_matrix(true_labels, predictions)

print("ğŸ“Š CONFUSION MATRIX:")
print("-"*70)
print(cm)
print()

# Confusion matrix gÃ¶rselleÅŸtirme
fig, ax = plt.subplots(figsize=(12, 10))

# Emotion labellarÄ±nÄ± kÄ±sa hale getir
emotion_names = [emotion_labels[i].split()[0] for i in range(6)]  # Sadece emoji'yi al

# Heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=emotion_names,
            yticklabels=emotion_names,
            cbar_kws={'label': 'Ã–rnek SayÄ±sÄ±'},
            linewidths=1, linecolor='gray',
            square=True, ax=ax)

ax.set_xlabel('Tahmin Edilen Duygu', fontsize=14, fontweight='bold', labelpad=10)
ax.set_ylabel('GerÃ§ek Duygu', fontsize=14, fontweight='bold', labelpad=10)
ax.set_title('ğŸ­ Confusion Matrix - BERT Emotion Classification', 
             fontsize=16, fontweight='bold', pad=20)

# Her hÃ¼creye yÃ¼zde ekle
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        percentage = cm[i, j] / cm[i].sum() * 100
        ax.text(j + 0.5, i + 0.7, f'({percentage:.1f}%)', 
               ha='center', va='center', fontsize=9, color='gray', fontweight='bold')

plt.tight_layout()
plt.show()

print()
print("="*70)
print("âœ… Confusion Matrix oluÅŸturuldu!")
print("="*70)
print()

# SÄ±nÄ±f bazÄ±nda doÄŸruluk hesapla
print("ğŸ“ˆ SINIF BAZINDA DOÄRULUK:")
print("-"*70)
for i in range(6):
    class_accuracy = cm[i, i] / cm[i].sum() * 100
    emotion_name = emotion_labels[i]
    print(f"{emotion_name:25s}: {cm[i, i]:4d}/{cm[i].sum():4d} ({class_accuracy:>5.2f}%)")

print()

# En Ã§ok karÄ±ÅŸtÄ±rÄ±lan durumlar
print("ğŸ” EN Ã‡OK KARIÅTIRILAN DURUMLAR:")
print("-"*70)
confusion_pairs = []
for i in range(6):
    for j in range(6):
        if i != j and cm[i, j] > 0:
            confusion_pairs.append((cm[i, j], i, j))

confusion_pairs.sort(reverse=True)

for count, true_idx, pred_idx in confusion_pairs[:5]:
    percentage = count / cm[true_idx].sum() * 100
    print(f"  â€¢ {emotion_labels[true_idx]:20s} â†’ {emotion_labels[pred_idx]:20s}: {count:3d} kez ({percentage:>5.2f}%)")

print()
print("="*70)
````

Classification report her duygu iÃ§in:
