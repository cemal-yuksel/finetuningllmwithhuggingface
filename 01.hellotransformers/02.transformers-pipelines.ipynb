````python
# üì¶ Adƒ±m 1: Gerekli K√ºt√ºphaneleri Y√ºkleyelim
# ============================================

# Transformers: Hugging Face'in NLP modelleri i√ßin ana k√ºt√ºphane
!pip install -U transformers

# SentencePiece: Metin tokenizasyonu i√ßin kullanƒ±lan k√ºt√ºphane
!pip install -U sentencepiece

# Sacremoses: Metin √∂n i≈üleme ve tokenizasyon i√ßin yardƒ±mcƒ± k√ºt√ºphane
!pip install -U sacremoses

print("\n‚úÖ T√ºm k√ºt√ºphaneler ba≈üarƒ±yla y√ºklendi!")
````

````python
# üìö Adƒ±m 2: K√ºt√ºphaneleri ƒ∞√ße Aktaralƒ±m
# ========================================

# Pipeline: √ñnceden eƒüitilmi≈ü modelleri kolayca kullanmamƒ±zƒ± saƒülar
from transformers import pipeline

# Pandas: Verileri tablo formatƒ±nda g√∂stermek i√ßin
import pandas as pd

print("‚úÖ K√ºt√ºphaneler ba≈üarƒ±yla i√ße aktarƒ±ldƒ±!")
````

````python
# ü§ñ Adƒ±m 3: Text Classification Pipeline'ƒ±nƒ± Olu≈üturalƒ±m
# =========================================================

# Duygu analizi i√ßin √∂zel olarak eƒüitilmi≈ü RoBERTa modelini y√ºkleyelim
# Bu model 27 farklƒ± duygu kategorisini tespit edebilir
classifier = pipeline("text-classification", model="SamLowe/roberta-base-go_emotions")

# Analiz edeceƒüimiz √∂rnek metin
# Bu metin bir ba≈üarƒ± ve heyecan ifadesi i√ßeriyor
text = "wow! we have come across this far"

# Metni modele g√∂nderelim ve duygu analizini yapalƒ±m
# Model, metindeki baskƒ±n duyguyu ve g√ºven skorunu d√∂nd√ºrecek
outputs = classifier(text)

# Sonu√ßlarƒ± g√ºzel bir tablo formatƒ±nda g√∂sterelim
# 'label': Tespit edilen duygu kategorisi
# 'score': Modelin bu tahmine olan g√ºven seviyesi (0-1 arasƒ±)
print("üéØ Duygu Analizi Sonu√ßlarƒ±:")
print("=" * 50)
pd.DataFrame(outputs)
````

````python
# üíº Adƒ±m 4: MIS (Management Information Systems) ile ƒ∞lgili √ñrnek C√ºmleler
# ============================================================================

# MIS konusunda farklƒ± uzunluklarda ƒ∞ngilizce c√ºmleler olu≈üturalƒ±m ve duygu analizini yapalƒ±m
# Model ƒ∞ngilizce metinleri daha iyi anlayacak ve daha doƒüru sonu√ßlar verecek

print("üìù MIS (MANAGEMENT INFORMATION SYSTEMS) SAMPLE SENTENCES - EMOTION ANALYSIS")
print("=" * 80)
print()

# üîπ SHORT SENTENCES (5-8 words)
# --------------------------------
short_sentences = [
    "The MIS course is very difficult.",
    "I completed the database project!",
    "The system analysis exam is approaching.",
    "ERP systems are amazing.",
    "SQL query is throwing an error."
]

print("üîπ SHORT SENTENCES:")
print("-" * 80)
for i, sentence in enumerate(short_sentences, 1):
    result = classifier(sentence)[0]
    print(f"{i}. '{sentence}'")
    print(f"   Emotion: {result['label']} | Confidence: {result['score']:.2%}")
    print()

print("\n")

# üî∏ MEDIUM-LENGTH SENTENCES (10-15 words)
# ------------------------------------------
medium_sentences = [
    "The MIS department is implementing a new enterprise resource planning system.",
    "We are analyzing customer behavior using data mining techniques and algorithms.",
    "Business intelligence tools have significantly accelerated our reporting processes and efficiency.",
    "Cloud computing solutions have substantially reduced the company's operational costs and expenses.",
    "I have serious concerns about system security, urgent measures are needed immediately."
]

print("üî∏ MEDIUM-LENGTH SENTENCES:")
print("-" * 80)
for i, sentence in enumerate(medium_sentences, 1):
    result = classifier(sentence)[0]
    print(f"{i}. '{sentence}'")
    print(f"   Emotion: {result['label']} | Confidence: {result['score']:.2%}")
    print()

print("\n")

# üî∂ LONG SENTENCES (20+ words)
# ------------------------------
long_sentences = [
    "The topics we learned in the Management Information Systems course, including enterprise resource planning, customer relationship management, and supply chain management, were truly very beneficial and I will definitely use them in my professional career.",
    
    "We are working to establish an integrated information system for all departments as part of our company's digital transformation project, but due to budget constraints and staff resistance, the process is progressing slower than expected, which is really frustrating and disappointing.",
    
    "By using data analytics and business intelligence tools, we can monitor the performance of our marketing campaigns in detail, and as a result, our customer satisfaction rates have increased by thirty-five percent in the last six months, which is truly an amazing achievement!",
    
    "Thanks to the training we received in information security and cybersecurity, we can better protect our company data, and employee awareness and compliance with security protocols is critically important, and we have made significant progress in this area.",
    
    "The new decision support system we have established provides great convenience to top management in making strategic decisions, and with real-time data analysis and predictive modeling features, we can foresee market trends in advance and gain a competitive advantage."
]

print("üî∂ LONG SENTENCES:")
print("-" * 80)
for i, sentence in enumerate(long_sentences, 1):
    result = classifier(sentence)[0]
    print(f"{i}. '{sentence[:80]}...'")  # Show first 80 characters
    print(f"   Emotion: {result['label']} | Confidence: {result['score']:.2%}")
    print()

print("\n" + "=" * 80)
print("‚úÖ Emotion analysis completed for all MIS sentences!")
````

````python
# üè∑Ô∏è Adƒ±m 5: Named Entity Recognition Pipeline'ƒ± Olu≈üturalƒ±m
# ===============================================================

# NER (Named Entity Recognition) pipeline'ƒ±nƒ± olu≈üturalƒ±m
# Bu model, metinden anahtar kelimeleri √ßƒ±karmak i√ßin √∂zel eƒüitilmi≈ütir

# aggregation_strategy="simple": Aynƒ± varlƒ±ƒüƒ±n farklƒ± tokenlarƒ±nƒ± birle≈ütirir
# √ñrneƒüin: "key" ve "phrase" ayrƒ± ayrƒ± deƒüil "keyphrase" olarak tanƒ±nƒ±r
ner_tagger = pipeline(
    "ner",                                                    # Task: Named Entity Recognition
    aggregation_strategy="simple",                           # Token'larƒ± birle≈ütir
    model="ml6team/keyphrase-extraction-kbir-inspec"        # Anahtar kelime √ßƒ±karma modeli
)

print("‚úÖ NER Pipeline ba≈üarƒ±yla olu≈üturuldu!")
print(f"üìå Model: ml6team/keyphrase-extraction-kbir-inspec")
print(f"üéØ G√∂rev: Keyphrase Extraction (Anahtar Kelime √áƒ±karma)")
````

````python
# üìù Adƒ±m 6: √ñrnek Metin √úzerinde Anahtar Kelime √áƒ±karma
# ==========================================================

# Analiz edeceƒüimiz teknik metin
# Bu metin, "keyphrase extraction" konusunu a√ßƒ±klƒ±yor
text = """Keyphrase extraction is a technique in text analysis where you extract the 
important keyphrases from a document. Thanks to these keyphrases humans can understand 
the content of a text very quickly and easily without reading it completely."""

print("üìÑ ANALƒ∞Z EDƒ∞LECEK METƒ∞N:")
print("=" * 80)
print(text)
print("=" * 80)
print()

# Metni NER modeline g√∂nderelim ve anahtar kelimeleri √ßƒ±karalƒ±m
# Model, metindeki en √∂nemli terimleri, konumlarƒ±nƒ± ve g√ºven skorlarƒ±nƒ± d√∂nd√ºrecek
outputs = ner_tagger(text)

print(f"‚úÖ Analiz tamamlandƒ±! {len(outputs)} anahtar kelime tespit edildi.")
print()

# Sonu√ßlarƒ± pandas DataFrame formatƒ±nda g√∂rselle≈ütirelim
# Her satƒ±r: bir anahtar kelime, varlƒ±k tipi, g√ºven skoru ve metin i√ßindeki konumu
print("üè∑Ô∏è TESPƒ∞T EDƒ∞LEN ANAHTAR KELƒ∞MELER:")
print("=" * 80)
df_results = pd.DataFrame(outputs)

# DataFrame'i daha okunabilir hale getirelim
if not df_results.empty:
    # Skor s√ºtununu y√ºzdeye √ßevirelim
    df_results['score'] = df_results['score'].apply(lambda x: f"{x:.2%}")
    print(df_results.to_string(index=False))
else:
    print("‚ö†Ô∏è Anahtar kelime bulunamadƒ±.")

print("=" * 80)

# Orijinal DataFrame'i g√∂sterelim (detaylƒ± analiz i√ßin)
pd.DataFrame(outputs)
````

````python
# üîç Adƒ±m 7: Daha Fazla √ñrnek - Farklƒ± Konularda Anahtar Kelime √áƒ±karma
# ========================================================================

# Farklƒ± alanlardaki metinlerden anahtar kelime √ßƒ±karalƒ±m
# Bu, modelin √ße≈üitli konulardaki performansƒ±nƒ± g√∂sterecek

print("üéØ √áE≈ûƒ∞TLƒ∞ KONULARDA ANAHTAR KELƒ∞ME √áƒ∞KARMA")
print("=" * 90)
print()

# ü§ñ Yapay Zeka konulu metin
ai_text = """Machine learning is a subset of artificial intelligence that focuses on 
algorithms and statistical models. Deep learning uses neural networks with multiple layers 
to learn hierarchical representations of data."""

print("ü§ñ YAPAY ZEKA METNƒ∞:")
print("-" * 90)
print(ai_text)
print()

ai_outputs = ner_tagger(ai_text)
print(f"‚úÖ {len(ai_outputs)} anahtar kelime bulundu:")
for item in ai_outputs:
    print(f"   ‚Ä¢ {item['word']} (Skor: {item['score']:.2%})")
print()
print()

# üíº ƒ∞≈ü D√ºnyasƒ± konulu metin
business_text = """Digital transformation is revolutionizing business processes and customer 
experience. Cloud computing and data analytics enable companies to make better strategic 
decisions and improve operational efficiency."""

print("üíº ƒ∞≈û D√úNYASI METNƒ∞:")
print("-" * 90)
print(business_text)
print()

business_outputs = ner_tagger(business_text)
print(f"‚úÖ {len(business_outputs)} anahtar kelime bulundu:")
for item in business_outputs:
    print(f"   ‚Ä¢ {item['word']} (Skor: {item['score']:.2%})")
print()
print()

# üè• Saƒülƒ±k konulu metin
health_text = """Healthcare informatics combines medical data with information technology. 
Electronic health records and telemedicine platforms improve patient care quality and 
enable remote diagnosis and treatment monitoring."""

print("üè• SAƒûLIK METNƒ∞:")
print("-" * 90)
print(health_text)
print()

health_outputs = ner_tagger(health_text)
print(f"‚úÖ {len(health_outputs)} anahtar kelime bulundu:")
for item in health_outputs:
    print(f"   ‚Ä¢ {item['word']} (Skor: {item['score']:.2%})")
print()

print("=" * 90)
print("üéä T√ºm metinler ba≈üarƒ±yla analiz edildi!")
print(f"üìä Toplam tespit edilen anahtar kelime sayƒ±sƒ±: {len(ai_outputs) + len(business_outputs) + len(health_outputs)}")
````

````python
# ‚ùì Adƒ±m 8: Question Answering Pipeline'ƒ± ve √ñrnek Metin
# =========================================================

# √ñnce analiz edeceƒüimiz context (baƒülam) metnini tanƒ±mlayalƒ±m
# Bu, Amazon'a g√∂nderilmi≈ü √∂rnek bir m√º≈üteri ≈üikayeti
text = """
Dear Amazon, last week I ordered an Optimus Prime action figure from your
online store in India. Unfortunately when I opened the package, I discovered to
my horror that I had been sent an action figure of Megatron instead!
"""

print("üìÑ CONTEXT (BAƒûLAM METNƒ∞):")
print("=" * 80)
print(text.strip())
print("=" * 80)
print()

# Question Answering pipeline'ƒ±nƒ± olu≈üturalƒ±m
# Bu pipeline, default olarak DistilBERT modelini kullanƒ±r (SQuAD veri seti ile eƒüitilmi≈ü)
reader = pipeline("question-answering")

print("‚úÖ Question Answering Pipeline ba≈üarƒ±yla olu≈üturuldu!")
print("ü§ñ Model: distilbert-base-cased-distilled-squad (default)")
print("üìö Eƒüitim: SQuAD (Stanford Question Answering Dataset)")
print()
````

````python
# üîç Adƒ±m 9: ƒ∞lk Soruyu Soralƒ±m ve Cevabƒ± Alalƒ±m
# ==================================================

# Sorumuzu tanƒ±mlayalƒ±m
question = "From where did I place the order?"

print("‚ùì SORU:")
print("-" * 80)
print(f"'{question}'")
print("-" * 80)
print()

# Soruyu modele g√∂nderelim
# Model, question ve context parametrelerini alƒ±r
# D√∂nen sonu√ß: cevap metni, ba≈ülangƒ±√ß/biti≈ü pozisyonlarƒ± ve g√ºven skoru
outputs = reader(question=question, context=text)

print("‚úÖ CEVAP BULUNDU!")
print("=" * 80)
print(f"üìç Cevap: '{outputs['answer']}'")
print(f"üéØ G√ºven Skoru: {outputs['score']:.2%}")
print(f"üìä Ba≈ülangƒ±√ß Pozisyonu: {outputs['start']}")
print(f"üìä Biti≈ü Pozisyonu: {outputs['end']}")
print("=" * 80)
print()

# Sonucu DataFrame formatƒ±nda g√∂rselle≈ütirelim
print("üìä DETAYLI SONU√á TABLOSU:")
result_df = pd.DataFrame([outputs])

# Skoru daha okunabilir hale getirelim
result_df_display = result_df.copy()
result_df_display['score'] = result_df_display['score'].apply(lambda x: f"{x:.2%}")
print(result_df_display.to_string(index=False))
print()

# Orijinal DataFrame'i d√∂nd√ºrelim
pd.DataFrame([outputs])
````

````python
# üí¨ Adƒ±m 10: Aynƒ± Metne Farklƒ± Sorular Soralƒ±m
# =================================================

# Aynƒ± context √ºzerinde birden fazla soru soralƒ±m
# Bu, modelin √ße≈üitli bilgileri √ßƒ±karma yeteneƒüini g√∂sterecek

print("üéØ √áOK SORULU SORU-CEVAP ANALƒ∞Zƒ∞")
print("=" * 90)
print()

# Farklƒ± sorular tanƒ±mlayalƒ±m
questions = [
    "From where did I place the order?",           # Nereden sipari≈ü verdim?
    "What did I order?",                           # Ne sipari≈ü verdim?
    "When did I place the order?",                 # Ne zaman sipari≈ü verdim?
    "What did I receive instead?",                 # Bunun yerine ne aldƒ±m?
    "What was the problem with the order?"         # Sipari≈üle ilgili sorun neydi?
]

# Her soru i√ßin cevap alalƒ±m ve sonu√ßlarƒ± saklayalƒ±m
all_results = []

for i, question in enumerate(questions, 1):
    # Soruyu modele g√∂nder
    answer = reader(question=question, context=text)
    
    # Sonucu kaydet
    all_results.append({
        'question': question,
        'answer': answer['answer'],
        'confidence': answer['score']
    })
    
    # Her soruyu ve cevabƒ±nƒ± ekrana yazdƒ±r
    print(f"‚ùì Soru {i}: {question}")
    print(f"‚úÖ Cevap: {answer['answer']}")
    print(f"üéØ G√ºven: {answer['score']:.2%}")
    print("-" * 90)
    print()

print("=" * 90)
print(f"‚úÖ Toplam {len(questions)} soru ba≈üarƒ±yla cevaplanƒ±ndƒ±!")
print()

# T√ºm sonu√ßlarƒ± DataFrame formatƒ±nda g√∂ster
print("üìä T√úM SORULAR VE CEVAPLAR:")
results_df = pd.DataFrame(all_results)
results_df['confidence'] = results_df['confidence'].apply(lambda x: f"{x:.2%}")
print(results_df.to_string(index=False))
print()

# Orijinal DataFrame'i d√∂nd√ºr
pd.DataFrame(all_results)
````

````python
# üìù Adƒ±m 13: Summarization Pipeline ve Basit √ñrnek
# ====================================================

# √ñzetlenecek kƒ±sa bir √∂rnek metin (Amazon ≈üikayeti)
text = """
Dear Amazon, last week I ordered an Optimus Prime action figure from your
online store in India. Unfortunately when I opened the package, I discovered to 
my horror that I had been sent an action figure of Megatron instead!
"""

print("üìÑ ORƒ∞Jƒ∞NAL METƒ∞N:")
print("=" * 80)
print(text.strip())
print("=" * 80)
print(f"üìè Karakter Sayƒ±sƒ±: {len(text)}")
print()

# Summarization pipeline'ƒ±nƒ± olu≈üturalƒ±m
# Bu pipeline, varsayƒ±lan olarak BART veya T5 gibi bir abstractive model kullanƒ±r
summarizer = pipeline("summarization")

print("‚úÖ Summarization Pipeline ba≈üarƒ±yla olu≈üturuldu!")
print("ü§ñ Model: sshleifer/distilbart-cnn-12-6 (default)")
print("üìö Yakla≈üƒ±m: Abstractive Summarization")
print()

# Metni √∂zetle
print("üîÑ Metin √∂zetleniyor...")
outputs = summarizer(text)

print("‚úÖ √ñZET OLU≈ûTURULDU!")
print("=" * 80)
print(outputs[0]['summary_text'])
print("=" * 80)
print(f"üìè √ñzet Karakter Sayƒ±sƒ±: {len(outputs[0]['summary_text'])}")
print(f"üìä Sƒ±kƒ±≈ütƒ±rma Oranƒ±: {(1 - len(outputs[0]['summary_text'])/len(text))*100:.1f}%")
print()

# Sonu√ßlarƒ± g√∂ster
print("üìã DETAYLI √áIKTI:")
outputs
````

````python
# üé® Adƒ±m 23: Farklƒ± Parametrelerle Metin √úretimi Deneyleri
# ============================================================

# Text generation parametrelerini ayarlayarak farklƒ± sonu√ßlar elde edelim
# Temperature, top_k, top_p gibi parametreler √ºretilen metnin yaratƒ±cƒ±lƒ±ƒüƒ±nƒ± etkiler

print("üé® FARKLI PARAMETRELERLE METƒ∞N √úRETƒ∞Mƒ∞ DENEYLERƒ∞")
print("=" * 90)
print()

# Ortak prompt
base_prompt = "In the field of artificial intelligence and machine learning"

print(f"üìå Temel Prompt: '{base_prompt}'")
print("=" * 90)
print()

# Deney 1: D√º≈ü√ºk temperature (daha deterministik, g√ºvenli)
print("üß™ DENEY 1: D√º≈ü√ºk Temperature (0.5) - Muhafazakar √úretim")
print("-" * 90)
output1 = generator(
    base_prompt,
    max_length=80,
    temperature=0.5,
    num_return_sequences=1
)
print(output1[0]['generated_text'])
print()
print()

# Deney 2: Y√ºksek temperature (daha yaratƒ±cƒ±, riskli)
print("üß™ DENEY 2: Y√ºksek Temperature (1.5) - Yaratƒ±cƒ± √úretim")
print("-" * 90)
output2 = generator(
    base_prompt,
    max_length=80,
    temperature=1.5,
    num_return_sequences=1
)
print(output2[0]['generated_text'])
print()
print()

# Deney 3: Top-k sampling (en iyi k kelime arasƒ±ndan se√ß)
print("üß™ DENEY 3: Top-k Sampling (k=50) - Kontroll√º √áe≈üitlilik")
print("-" * 90)
output3 = generator(
    base_prompt,
    max_length=80,
    top_k=50,
    num_return_sequences=1
)
print(output3[0]['generated_text'])
print()
print()

# Deney 4: √áoklu metin √ºretimi (aynƒ± prompt'tan farklƒ± versiyonlar)
print("üß™ DENEY 4: √áoklu Metin √úretimi (3 Farklƒ± Versiyon)")
print("-" * 90)
outputs_multiple = generator(
    base_prompt,
    max_length=60,
    num_return_sequences=3,
    temperature=0.8
)

for idx, output in enumerate(outputs_multiple, 1):
    print(f"üìÑ Versiyon {idx}:")
    print(output['generated_text'])
    print()

print("=" * 90)
print("‚úÖ T√ºm deneyler tamamlandƒ±!")
print()
print("üí° G√ñZLEMLERƒ∞Mƒ∞Z:")
print("   ‚Ä¢ D√º≈ü√ºk temperature: Daha tutarlƒ± ve tahmin edilebilir metinler")
print("   ‚Ä¢ Y√ºksek temperature: Daha yaratƒ±cƒ± ama bazen mantƒ±ksƒ±z metinler")
print("   ‚Ä¢ Top-k sampling: √áe≈üitlilik ve kalite dengesi")
print("   ‚Ä¢ √áoklu √ºretim: Aynƒ± prompt'tan farklƒ± perspektifler")
````

````python
# üêß Adƒ±m 24: Basit G√∂r√ºnt√º Sƒ±nƒ±flandƒ±rma - Penguen Fotoƒürafƒ±
# URL'den bir penguen g√∂r√ºnt√ºs√º √ßekerek sƒ±nƒ±flandƒ±rma yapƒ±yoruz

from PIL import Image
import requests
from transformers import pipeline

# üîó ƒ∞nternetten penguen fotoƒürafƒ± URL'i
url = "https://res.cloudinary.com/dk-find-out/image/upload/q_80,w_1920,f_auto/DCTM_Penguin_UK_DK_AL697473_RGB_PNG_namnse.jpg"

# üì• G√∂r√ºnt√ºy√º indir ve PIL Image formatƒ±na √ßevir
print("üîÑ G√∂r√ºnt√º indiriliyor...")
image = Image.open(requests.get(url, stream=True).raw)
print(f"‚úÖ G√∂r√ºnt√º ba≈üarƒ±yla y√ºklendi! Boyut: {image.size}")

# ü§ñ Vision Transformer modeli ile image classification pipeline olu≈ütur
# Default olarak Google'ƒ±n ViT (Vision Transformer) modeli kullanƒ±lƒ±r
print("\nüöÄ Image Classification modeli y√ºkleniyor...")
classifier = pipeline("image-classification")

# üéØ G√∂r√ºnt√ºy√º sƒ±nƒ±flandƒ±r
print("\nüîç G√∂r√ºnt√º analiz ediliyor...\n")
outputs = classifier(image)

# üìä Sonu√ßlarƒ± g√∂ster
print("=" * 60)
print("üèÜ EN Y√úKSEK 5 TAHMƒ∞N")
print("=" * 60)
for idx, output in enumerate(outputs, 1):
    label = output['label']
    score = output['score']
    print(f"{idx}. {label:30s} - G√ºven: {score:.2%}")
print("=" * 60)
````

````python
# üë® Adƒ±m 25: √ñzel Model ile Ya≈ü Tahmini - ƒ∞nsan Y√ºz√º Sƒ±nƒ±flandƒ±rmasƒ±
# Farklƒ± bir URL ve √∂zel bir model (ya≈ü tahmin modeli) kullanƒ±yoruz

# üîó ƒ∞nternetten insan fotoƒürafƒ± URL'i
url = "https://img.freepik.com/free-photo/young-bearded-man-with-striped-shirt_273609-5677.jpg"

# üì• G√∂r√ºnt√ºy√º indir
print("üîÑ ƒ∞nsan fotoƒürafƒ± indiriliyor...")
image = Image.open(requests.get(url, stream=True).raw)
print(f"‚úÖ G√∂r√ºnt√º ba≈üarƒ±yla y√ºklendi! Boyut: {image.size}")

# üß† √ñzel bir model belirt: Ya≈ü tahmin modeli
# Bu model √∂zellikle insan y√ºzlerinden ya≈ü tahmini yapmak i√ßin eƒüitilmi≈ü
print("\nüöÄ Ya≈ü tahmin modeli y√ºkleniyor (nateraw/vit-age-classifier)...")
classifier = pipeline("image-classification", model="nateraw/vit-age-classifier")

# üéØ G√∂r√ºnt√ºy√º sƒ±nƒ±flandƒ±r ve ya≈ü grubunu tahmin et
print("\nüîç Ya≈ü analizi yapƒ±lƒ±yor...\n")
outputs = classifier(image)

# üìä Sonu√ßlarƒ± g√∂ster
print("=" * 60)
print("üë§ YA≈û GRUBU TAHMƒ∞NLERƒ∞")
print("=" * 60)
for idx, output in enumerate(outputs, 1):
    age_group = output['label']
    confidence = output['score']
    print(f"{idx}. Ya≈ü Grubu: {age_group:15s} - Olasƒ±lƒ±k: {confidence:.2%}")
print("=" * 60)
print(f"\nüéØ En olasƒ± ya≈ü grubu: {outputs[0]['label']} ({outputs[0]['score']:.2%} g√ºven)")
print("=" * 60)
````
