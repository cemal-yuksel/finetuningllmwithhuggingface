````python
# ğŸ² Tekrarlanabilirlik iÃ§in Seed DeÄŸerini Sabitleme

def set_seed(seed_value=42):
    """
    TÃ¼m rastgele sayÄ± Ã¼reticilerinin seed deÄŸerini sabitler.

    Args:
        seed_value (int): KullanÄ±lacak seed deÄŸeri (varsayÄ±lan: 42)

    Not:
        42 deÄŸeri ML topluluÄŸunda popÃ¼ler bir seed deÄŸeridir.
        ("The Hitchhiker's Guide to the Galaxy" kitabÄ±ndan esinlenilmiÅŸtir ğŸ˜Š)
    """
    # Python'un random modÃ¼lÃ¼
    random.seed(seed_value)

    # NumPy random seed
    np.random.seed(seed_value)

    # PyTorch random seed (CPU)
    torch.manual_seed(seed_value)

    # PyTorch random seed (GPU)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed_value)
        torch.cuda.manual_seed_all(seed_value)  # Multi-GPU iÃ§in

        # CuDNN'i deterministik yap (GPU hesaplamalarÄ±nÄ± sabitle)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

    # Python hash seed
    os.environ['PYTHONHASHSEED'] = str(seed_value)

    print(f"âœ… Seed deÄŸeri {seed_value} olarak sabitlendi!")
    print("ğŸ”’ TÃ¼m rastgele iÅŸlemler artÄ±k tekrarlanabilir.")

# Seed'i ayarla
SEED = 42
set_seed(SEED)
````

## 4.2 ğŸ§¹ Eksik Veri Analizi

## 4.3 âš–ï¸ SÄ±nÄ±f DengesizliÄŸi Analizi

ML modellerinde **sÄ±nÄ±f dengesizliÄŸi** (class imbalance) Ã¶nemli bir sorundur. GerÃ§ek ve sahte haber sayÄ±larÄ±nÄ±n dengeli olup olmadÄ±ÄŸÄ±nÄ± kontrol edelim.

````python
# âš–ï¸ SÄ±nÄ±f DaÄŸÄ±lÄ±mÄ± Analizi

if 'label' in df.columns:
    print("=" * 80)
    print("âš–ï¸  SINIF DAÄILIMI ANALÄ°ZÄ°")
    print("=" * 80)

    # SÄ±nÄ±f sayÄ±larÄ±
    class_counts = df['label'].value_counts()
    class_percentages = df['label'].value_counts(normalize=True) * 100

    print("\nğŸ“Š SÄ±nÄ±f SayÄ±larÄ±:")
    print(f"   GerÃ§ek Haber (0): {class_counts.get(0, 0):,} adet ({class_percentages.get(0, 0):.2f}%)")
    print(f"   Sahte Haber (1):  {class_counts.get(1, 0):,} adet ({class_percentages.get(1, 0):.2f}%)")

    # Dengesizlik oranÄ± hesapla
    if len(class_counts) == 2:
        imbalance_ratio = max(class_counts) / min(class_counts)
        print(f"\nğŸ“ˆ Dengesizlik OranÄ±: {imbalance_ratio:.2f}:1")

        if imbalance_ratio > 1.5:
            print("âš ï¸  SÄ±nÄ±flar dengesiz! Class weighting veya oversampling dÃ¼ÅŸÃ¼nÃ¼lebilir.")
        else:
            print("âœ… SÄ±nÄ±flar nispeten dengeli gÃ¶rÃ¼nÃ¼yor!")

    # GÃ¶rselleÅŸtirme - 2 farklÄ± grafik
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # 1. Bar Chart (Pastel renkler)
    colors_bar = ['#98FB98', '#FFB6C1']  # YeÅŸil ve pembe tonlarÄ±
    class_counts.plot(kind='bar', ax=axes[0], color=colors_bar, edgecolor='black', linewidth=1.5)
    axes[0].set_title('ğŸ“Š SÄ±nÄ±f DaÄŸÄ±lÄ±mÄ± (Bar Chart)', fontsize=14, fontweight='bold', pad=15)
    axes[0].set_xlabel('SÄ±nÄ±f (0=GerÃ§ek, 1=Sahte)', fontsize=12, fontweight='bold')
    axes[0].set_ylabel('KayÄ±t SayÄ±sÄ±', fontsize=12, fontweight='bold')
    axes[0].set_xticklabels(['GerÃ§ek (0)', 'Sahte (1)'], rotation=0)
    axes[0].grid(axis='y', alpha=0.3, linestyle='--')

    # DeÄŸerleri bar'larÄ±n Ã¼stÃ¼ne yaz
    for i, v in enumerate(class_counts):
        axes[0].text(i, v + 100, f'{v:,}', ha='center', va='bottom', fontweight='bold', fontsize=11)

    # 2. Pie Chart (Pastel renkler)
    colors_pie = ['#87CEEB', '#FFB6C1']  # Mavi ve pembe tonlarÄ±
    labels_pie = [f'GerÃ§ek\n{class_percentages.get(0, 0):.1f}%', f'Sahte\n{class_percentages.get(1, 0):.1f}%']

    axes[1].pie(class_counts, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%',
                startangle=90, textprops={'fontsize': 12, 'fontweight': 'bold'},
                wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})
    axes[1].set_title('ğŸ¥§ SÄ±nÄ±f OranlarÄ± (Pie Chart)', fontsize=14, fontweight='bold', pad=15)

    plt.tight_layout()
    plt.show()

    print("\nğŸ’¼ YBS Ä°Ã§gÃ¶rÃ¼sÃ¼:")
    print("   SÄ±nÄ±f dengesizliÄŸi, iÅŸ sÃ¼reÃ§lerinde 'olay riski' ile benzerdir.")
    print("   Ã–rnek: Kredi baÅŸvurularÄ±nda %95 onay, %5 ret - nadir durumu tespit etmek zor!")

    print("=" * 80)
else:
    print("âš ï¸  'label' sÃ¼tunu bulunamadÄ±!")
````

## 4.5 ğŸ’¬ Kelime Frekans Analizi ve Word Cloud

En sÄ±k kullanÄ±lan kelimeleri analiz edelim ve gÃ¶rselleÅŸtirelim.

````python
# ğŸ’¬ Kelime Frekans Analizi

from collections import Counter
import re

if 'title' in df.columns:
    print("=" * 80)
    print("ğŸ’¬ KELÄ°ME FREKANS ANALÄ°ZÄ°")
    print("=" * 80)

    # TÃ¼m baÅŸlÄ±klarÄ± birleÅŸtir ve kelimelere ayÄ±r
    all_titles = ' '.join(df['title'].fillna('').astype(str).values)

    # Metni temizle ve kelimelere ayÄ±r (kÃ¼Ã§Ã¼k harfe Ã§evir, noktalama iÅŸaretlerini kaldÄ±r)
    words = re.findall(r'\b[a-zA-Z]{3,}\b', all_titles.lower())  # En az 3 harfli kelimeler

    # YaygÄ±n Ä°ngilizce stop words (durma kelimeleri)
    stop_words = {'the', 'and', 'for', 'that', 'with', 'this', 'from', 'are', 'was',
                  'has', 'have', 'will', 'been', 'said', 'says', 'can', 'they', 'you',
                  'not', 'more', 'but', 'who', 'what', 'when', 'how', 'where', 'why',
                  'one', 'two', 'their', 'about', 'after', 'all', 'also', 'than'}

    # Stop words'leri filtrele
    words_filtered = [word for word in words if word not in stop_words]

    # FrekanslarÄ± hesapla
    word_freq = Counter(words_filtered)
    most_common = word_freq.most_common(20)

    print(f"\nğŸ“Š En SÄ±k KullanÄ±lan 20 Kelime:")
    for i, (word, count) in enumerate(most_common, 1):
        print(f"   {i:2d}. {word:15s} : {count:4d} kez")

    # GÃ¶rselleÅŸtirme - Bar Chart
    plt.figure(figsize=(12, 6))

    words_list = [item[0] for item in most_common[:15]]
    counts_list = [item[1] for item in most_common[:15]]

    # Pastel gÃ¶kkuÅŸaÄŸÄ± renkleri
    colors = plt.cm.Pastel1(range(len(words_list)))

    bars = plt.bar(words_list, counts_list, color=colors, edgecolor='black', linewidth=1.5)

    # Her bar'Ä±n Ã¼stÃ¼ne deÄŸeri yaz
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{int(height)}',
                ha='center', va='bottom', fontweight='bold', fontsize=10)

    plt.title('ğŸ’¬ En SÄ±k KullanÄ±lan 15 Kelime', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Kelimeler', fontsize=12, fontweight='bold')
    plt.ylabel('Frekans', fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3, linestyle='--')
    plt.tight_layout()
    plt.show()

    print("\nğŸ’¼ YBS Ä°Ã§gÃ¶rÃ¼sÃ¼:")
    print("   Kelime frekanslarÄ±, metin madenciliÄŸi (text mining) ve sentiment analizi")
    print("   iÃ§in Ã¶nemlidir. Ä°ÅŸletmelerde mÃ¼ÅŸteri yorumlarÄ±nÄ± analiz ederken benzer")
    print("   teknikleri kullanÄ±rÄ±z!")

    print("=" * 80)
else:
    print("âš ï¸  'title' sÃ¼tunu bulunamadÄ±!")
````

<div style="background: linear-gradient(135deg, #F0E5FF 0%, #DDA0DD 100%); padding: 20px; border-radius: 10px; border-left: 5px solid #9370DB;">
    <h4>ğŸ“ Metin UzunluÄŸu ve Token Limitleri</h4>
    <p><strong>Neden Ã–nemli:</strong> BERT modelleri token limitine sahiptir (genelde 512).</p>
    <p><strong>Token Nedir:</strong> Kelimelerin veya alt-kelimelerin sayÄ±sal temsilleri.</p>
    <p><strong>Kural:</strong></p>
    <ul>
        <li>1 Ä°ngilizce kelime â‰ˆ 1-2 token</li>
        <li>1 token â‰ˆ 4 karakter (ortalama)</li>
    </ul>
    <p><strong>Ã–rnek:</strong></p>
    <ul>
        <li>"Hello world" â†’ 2 token</li>
        <li>"Unbelievable" â†’ 3 token (un-believ-able)</li>
    </ul>
    <p><strong>Uzun Metinler Ä°Ã§in:</strong></p>

````python
# âœ‚ï¸ Train-Validation-Test Split

print("=" * 80)
print("âœ‚ï¸  VERÄ° BÃ–LME Ä°ÅLEMÄ°")
print("=" * 80)

# Parametreler
TRAIN_SIZE = 0.70  # %70 eÄŸitim
VAL_SIZE = 0.15    # %15 doÄŸrulama
TEST_SIZE = 0.15   # %15 test
RANDOM_STATE = 42  # Tekrarlanabilirlik iÃ§in

print(f"\nğŸ“Š BÃ¶lme OranlarÄ±:")
print(f"   ğŸ“ Train:      {TRAIN_SIZE*100:.0f}%")
print(f"   âœ… Validation: {VAL_SIZE*100:.0f}%")
print(f"   ğŸ§ª Test:       {TEST_SIZE*100:.0f}%")

# Ä°lk bÃ¶lme: Train ve Temp (Val+Test)
train_df, temp_df = train_test_split(
    df_clean,
    test_size=(VAL_SIZE + TEST_SIZE),  # 0.30
    random_state=RANDOM_STATE,
    stratify=df_clean['label']  # SÄ±nÄ±f dengesini koru (stratified split)
)

# Ä°kinci bÃ¶lme: Validation ve Test
val_df, test_df = train_test_split(
    temp_df,
    test_size=TEST_SIZE/(VAL_SIZE + TEST_SIZE),  # 0.5 (Ã§Ã¼nkÃ¼ temp'in yarÄ±sÄ±)
    random_state=RANDOM_STATE,
    stratify=temp_df['label']  # SÄ±nÄ±f dengesini koru
)

# SonuÃ§larÄ± gÃ¶ster
print(f"\nğŸ“ˆ BÃ¶lme SonuÃ§larÄ±:")
print(f"   ğŸ“š Train Set:      {len(train_df):,} kayÄ±t ({len(train_df)/len(df_clean)*100:.1f}%)")
print(f"   âœ… Validation Set: {len(val_df):,} kayÄ±t ({len(val_df)/len(df_clean)*100:.1f}%)")
print(f"   ğŸ§ª Test Set:       {len(test_df):,} kayÄ±t ({len(test_df)/len(df_clean)*100:.1f}%)")
print(f"   ğŸ“¦ TOPLAM:         {len(train_df) + len(val_df) + len(test_df):,} kayÄ±t")

# Her set'teki sÄ±nÄ±f daÄŸÄ±lÄ±mÄ±nÄ± kontrol et
print(f"\nâš–ï¸  SÄ±nÄ±f DaÄŸÄ±lÄ±mÄ± KontrolÃ¼:")

for name, subset in [('Train', train_df), ('Validation', val_df), ('Test', test_df)]:
    class_dist = subset['label'].value_counts(normalize=True) * 100
    print(f"\n   {name} Set:")
    print(f"      Real (0): {class_dist.get(0, 0):.1f}%")
    print(f"      Fake (1): {class_dist.get(1, 0):.1f}%")

# GÃ¶rselleÅŸtirme
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1. Set boyutlarÄ± (Bar Chart)
set_sizes = [len(train_df), len(val_df), len(test_df)]
set_names = ['Train\n(70%)', 'Validation\n(15%)', 'Test\n(15%)']
colors = ['#98FB98', '#87CEEB', '#FFB6C1']

bars = axes[0].bar(set_names, set_sizes, color=colors, edgecolor='black', linewidth=2)
axes[0].set_title('ğŸ“Š Veri Seti BoyutlarÄ±', fontsize=14, fontweight='bold', pad=15)
axes[0].set_ylabel('KayÄ±t SayÄ±sÄ±', fontsize=12, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3, linestyle='--')

# DeÄŸerleri bar'larÄ±n Ã¼stÃ¼ne yaz
for bar, size in zip(bars, set_sizes):
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2., height,
                f'{size:,}',
                ha='center', va='bottom', fontweight='bold', fontsize=11)

# 2. SÄ±nÄ±f daÄŸÄ±lÄ±mÄ± karÅŸÄ±laÅŸtÄ±rmasÄ± (Grouped Bar Chart)
train_dist = train_df['label'].value_counts(normalize=True) * 100
val_dist = val_df['label'].value_counts(normalize=True) * 100
test_dist = test_df['label'].value_counts(normalize=True) * 100

x = np.arange(2)  # 2 sÄ±nÄ±f var
width = 0.25

axes[1].bar(x - width, [train_dist.get(0, 0), train_dist.get(1, 0)], width,
           label='Train', color='#98FB98', edgecolor='black')
axes[1].bar(x, [val_dist.get(0, 0), val_dist.get(1, 0)], width,
           label='Validation', color='#87CEEB', edgecolor='black')
axes[1].bar(x + width, [test_dist.get(0, 0), test_dist.get(1, 0)], width,
           label='Test', color='#FFB6C1', edgecolor='black')

axes[1].set_title('âš–ï¸ Set\'lerdeki SÄ±nÄ±f DaÄŸÄ±lÄ±mÄ±', fontsize=14, fontweight='bold', pad=15)
axes[1].set_ylabel('Oran (%)', fontsize=12, fontweight='bold')
axes[1].set_xticks(x)
axes[1].set_xticklabels(['Real (0)', 'Fake (1)'])
axes[1].legend(fontsize=10)
axes[1].grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

print("\nğŸ’¡ Stratified Split KullanÄ±ldÄ±:")
print("   Her set'te sÄ±nÄ±f oranlarÄ± korundu (dengeli bÃ¶lme)")
print("   Bu sayede model tÃ¼m set'lerde adil bir ÅŸekilde test edilebilir!")

print("=" * 80)
````

<div style="background: linear-gradient(135deg, #FFFACD 0%, #FFD700 100%); padding: 20px; border-radius: 10px; border-left: 5px solid #DAA520;">
    <h4>âš¡ Batch Processing</h4>
    <p><strong>Problem:</strong> FarklÄ± uzunluktaki metinleri aynÄ± anda iÅŸleyemezsiniz!</p>
    <p><strong>Ã‡Ã¶zÃ¼m:</strong></p>
    <ul>
        <li><strong>Padding:</strong> KÄ±sa metinlere [PAD] token ekleyerek uzat</li>
        <li><strong>Truncation:</strong> Uzun metinlerden fazla kÄ±smÄ± kes</li>
        <li><strong>Max Length:</strong> TÃ¼m metinleri aynÄ± uzunluÄŸa getir (Ã¶rn: 128)</li>
    </ul>
    <p><strong>YBS'de:</strong> VeritabanÄ± tablolarÄ±nda sabit uzunluklu VARCHAR alanlar gibi!</p>

````python
# âš™ï¸ Padding ve Truncation Ã–rneÄŸi

print("=" * 80)
print("âš™ï¸  PADDING & TRUNCATION Ã–RNEÄÄ°")
print("=" * 80)

# FarklÄ± uzunlukta metinler
short_text = "AI is great"
medium_text = "Artificial intelligence is transforming the business world"
long_text = "Artificial intelligence and machine learning are revolutionizing how businesses operate, making processes more efficient and enabling data-driven decision making across all departments"

texts = [short_text, medium_text, long_text]

print("\nğŸ“ Ã–rnek Metinler:")
for i, text in enumerate(texts, 1):
    print(f"   {i}. '{text[:50]}...' ({len(text)} karakter)")

# Tokenize et (padding VE truncation ile)
MAX_LENGTH = 20  # Demo iÃ§in kÄ±sa tuttuk

print(f"\nğŸ”§ Parametreler:")
print(f"   Max Length: {MAX_LENGTH}")
print(f"   Padding: 'max_length'")
print(f"   Truncation: True")

# Batch encoding
encoded_batch = tokenizer(
    texts,
    padding='max_length',  # TÃ¼m metinleri max_length'e getir
    truncation=True,       # Uzun metinleri kes
    max_length=MAX_LENGTH,
    return_tensors='pt'    # PyTorch tensor olarak dÃ¶ndÃ¼r
)

print(f"\nğŸ“Š Tokenization SonuÃ§larÄ±:")
print(f"   Input IDs Shape: {encoded_batch['input_ids'].shape}")
print(f"   (batch_size={len(texts)}, max_length={MAX_LENGTH})")

# Her metin iÃ§in detaylarÄ± gÃ¶ster
for i, text in enumerate(texts):
    print(f"\n   --- Metin {i+1} ---")
    print(f"   Orijinal: '{text[:40]}...'")

    # Token'larÄ± gÃ¶ster
    tokens = tokenizer.convert_ids_to_tokens(encoded_batch['input_ids'][i])
    print(f"   Token'lar: {tokens[:10]}...")

    # GerÃ§ek token sayÄ±sÄ± (padding hariÃ§)
    real_tokens = (encoded_batch['attention_mask'][i] == 1).sum().item()
    padding_count = MAX_LENGTH - real_tokens

    print(f"   GerÃ§ek Token: {real_tokens}")
    print(f"   Padding:      {padding_count}")

    if len(tokenizer.tokenize(text)) > MAX_LENGTH - 2:  # -2 for [CLS] and [SEP]
        print(f"   âš ï¸  Truncation uygulandÄ±!")

# GÃ¶rselleÅŸtirme
print(f"\nğŸ“Š Attention Mask GÃ¶rselleÅŸtirmesi:")
for i in range(len(texts)):
    mask_visual = ''.join(['â–ˆ' if m == 1 else 'â–‘' for m in encoded_batch['attention_mask'][i].tolist()])
    print(f"   Metin {i+1}: {mask_visual}")
    print(f"           (â–ˆ = gerÃ§ek token, â–‘ = padding)")

print("\nğŸ’¡ YBS Ä°Ã§gÃ¶rÃ¼sÃ¼:")
print("   Padding, veritabanlarÄ±nda VARCHAR alanlarÄ±na benzer.")
print("   Truncation ise veri validasyonunda maksimum uzunluk kontrolÃ¼ne benzer!")

print("=" * 80)
````

````python
# ğŸ”§ Tokenization Fonksiyonu

def tokenize_function(examples):
    """
    Batch tokenization fonksiyonu

    Args:
        examples: Bir batch veri (dict formatÄ±nda, 'title' key'i iÃ§erir)

    Returns:
        Tokenize edilmiÅŸ batch (input_ids, attention_mask)
    """
    # Batch iÅŸleme - tÃ¼m metinleri bir anda tokenize et
    return tokenizer(
        examples['title'],      # Metin sÃ¼tunu
        padding='max_length',   # TÃ¼m Ã¶rnekleri aynÄ± uzunluÄŸa getir
        truncation=True,        # Uzun metinleri kes
        max_length=128,         # Maksimum 128 token (DistilBERT iÃ§in uygun)
        return_tensors=None     # map() iÃ§in tensÃ¶re Ã§evirme (otomatik olacak)
    )

print("âœ… Tokenization fonksiyonu hazÄ±r!")
print("\nğŸ“‹ Fonksiyon DetaylarÄ±:")
print(f"   - Max Length: 128 token")
print(f"   - Padding: max_length (tÃ¼m batch'te uniform uzunluk)")
print(f"   - Truncation: Aktif (128 token'dan uzun metinler kesilecek)")

# Test edelim
sample = {"title": ["Test news 1", "Test news 2"]}
test_output = tokenize_function(sample)
print(f"\nğŸ§ª Test Ã§Ä±ktÄ±sÄ±:")
print(f"   Keys: {list(test_output.keys())}")
print(f"   Input IDs shape: {len(test_output['input_ids'])} x {len(test_output['input_ids'][0])}")
````

````python
# ğŸ·ï¸ Label Mapping TanÄ±mlama

# Veri setimizde: 0 = Real, 1 = Fake
label2id = {
    "REAL": 0,   # GerÃ§ek haber
    "FAKE": 1    # Sahte haber
}

id2label = {
    0: "REAL",   # Model 0 Ã§Ä±ktÄ±sÄ± verirse â†’ Real
    1: "FAKE"    # Model 1 Ã§Ä±ktÄ±sÄ± verirse â†’ Fake
}

print("ğŸ·ï¸  LABEL MAPPING")
print("=" * 60)
print("\nğŸ“‹ Label2ID (Ä°nsan â†’ Model):")
for label, idx in label2id.items():
    print(f"   '{label}' â†’ {idx}")

print("\nğŸ“‹ ID2Label (Model â†’ Ä°nsan):")
for idx, label in id2label.items():
    print(f"   {idx} â†’ '{label}'")

print("\nğŸ’¡ YBS Ä°Ã§gÃ¶rÃ¼sÃ¼:")
print("   Bu mapping, veritabanÄ±nda ENUM tiplerinin ID'lerle")
print("   eÅŸleÅŸtirilmesine benzer!")
print("   Ã–rnek: 'ACTIVE' â†’ 1, 'INACTIVE' â†’ 0")
print("=" * 60)
````

### ğŸ“Š 8.2 Evaluation Metrics TanÄ±mlama

Model performansÄ±nÄ± Ã¶lÃ§mek iÃ§in metric fonksiyonlarÄ± tanÄ±mlÄ±yoruz.

````python
# ğŸ“Š Evaluation Metrics Fonksiyonu

from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import evaluate

# HuggingFace evaluate kÃ¼tÃ¼phanesi ile metric yÃ¼kle
accuracy_metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    """
    Evaluation sÄ±rasÄ±nda metrikleri hesaplayan fonksiyon

    Args:
        eval_pred: Tuple (predictions, labels)
            - predictions: Model Ã§Ä±ktÄ±larÄ± (logits)
            - labels: GerÃ§ek label'lar

    Returns:
        dict: Metrik sonuÃ§larÄ±
    """
    # Predictions ve labels'Ä± ayÄ±kla
    logits, labels = eval_pred

    # Logits'ten tahmin edilen sÄ±nÄ±flarÄ± al (argmax)
    predictions = np.argmax(logits, axis=-1)

    # Accuracy hesapla (HuggingFace)
    accuracy = accuracy_metric.compute(predictions=predictions, references=labels)

    # Precision, Recall, F1 hesapla (Scikit-learn)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels,
        predictions,
        average='binary',  # Binary classification iÃ§in
        pos_label=1        # Fake (1) pozitif sÄ±nÄ±f
    )

    # SonuÃ§larÄ± dÃ¶ndÃ¼r
    return {
        'accuracy': accuracy['accuracy'],
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

print("âœ… Evaluation metrics fonksiyonu tanÄ±mlandÄ±!")
print("\nğŸ“Š Hesaplanacak Metrikler:")
print("   â€¢ Accuracy: Genel doÄŸruluk oranÄ±")
print("   â€¢ F1 Score: Precision ve Recall'un harmonik ortalamasÄ±")
print("   â€¢ Precision: Fake dediÄŸimiz haberlerin ne kadarÄ± gerÃ§ekten fake?")
print("   â€¢ Recall: TÃ¼m fake haberlerin ne kadarÄ±nÄ± yakaladÄ±k?")

print("\nğŸ’¡ YBS Ä°Ã§gÃ¶rÃ¼sÃ¼:")
print("   Accuracy = Genel baÅŸarÄ± oranÄ± (KPI)")
print("   Precision = YanlÄ±ÅŸ alarm oranÄ± (MÃ¼ÅŸteri memnuniyeti)")
print("   Recall = Yakalama oranÄ± (Risk yÃ¶netimi)")
print("   F1 = Dengeli performans Ã¶lÃ§Ã¼mÃ¼ (Overall score)")
````

### ğŸ”§ 8.3 Training Arguments YapÄ±landÄ±rmasÄ±

EÄŸitim sÃ¼recinin tÃ¼m parametrelerini tanÄ±mlÄ±yoruz.

````python
# âš™ï¸ Training Arguments TanÄ±mlama

from transformers import TrainingArguments

# Ã‡Ä±ktÄ± dizinini daha Ã¶nce tanÄ±mladÄ±ÄŸÄ±mÄ±z klasÃ¶rlerden al
output_dir = OUTPUT_DIRS['models']  # Google Drive veya local path
logging_dir = OUTPUT_DIRS['logs']   # Log klasÃ¶rÃ¼

print("âš™ï¸  TRAINING ARGUMENTS YAPISI")
print("=" * 80)
print(f"\nğŸ“ Ã‡Ä±ktÄ± KlasÃ¶rleri:")
print(f"   Model: {output_dir}")
print(f"   Logs: {logging_dir}")

# Training arguments oluÅŸtur
training_args = TrainingArguments(
    # === TEMEL AYARLAR ===
    output_dir=output_dir,                    # Model kayÄ±t dizini (Google Drive)
    overwrite_output_dir=True,                # Var olan modelin Ã¼zerine yaz

    # === EÄÄ°TÄ°M PARAMETRELERÄ° ===
    num_train_epochs=3,                       # 3 epoch eÄŸitim (daha fazla = daha iyi, ama overfit riski)
    per_device_train_batch_size=16,           # Her GPU/CPU iÃ§in batch size
    per_device_eval_batch_size=32,            # Evaluation batch size (daha bÃ¼yÃ¼k olabilir)

    # === LEARNING RATE ===
    learning_rate=2e-5,                       # Learning rate (2e-5 = 0.00002, BERT iÃ§in ideal)
    weight_decay=0.01,                        # Regularization (overfitting Ã¶nler)
    warmup_steps=500,                         # Ä°lk 500 adÄ±mda learning rate'i artÄ±r

    # === EVALUATION ===
    eval_strategy="steps",                    # Her N adÄ±mda evaluate et
    eval_steps=500,                           # Her 500 adÄ±mda evaluation yap
    save_strategy="steps",                    # Her N adÄ±mda checkpoint kaydet
    save_steps=500,                           # Her 500 adÄ±mda kaydet
    save_total_limit=2,                       # En fazla 2 checkpoint tut (disk tasarrufu)

    # === LOGGÄ°NG ===
    logging_dir=logging_dir,                  # TensorBoard loglarÄ± (Google Drive)
    logging_steps=100,                        # Her 100 adÄ±mda log

    # === PERFORMANCE ===
    load_best_model_at_end=True,              # EÄŸitim sonunda en iyi modeli yÃ¼kle
    metric_for_best_model="f1",               # En iyi model: En yÃ¼ksek F1 score
    greater_is_better=True,                   # F1 iÃ§in: YÃ¼ksek = iyi

    # === DIÄER ===
    push_to_hub=False,                        # HuggingFace Hub'a yÃ¼kleme (ÅŸimdilik hayÄ±r)
    report_to="none",                         # Raporlama platformu (none = sadece local)
    seed=42,                                  # Reproducibility
)

print("\nâœ… Training arguments yapÄ±landÄ±rÄ±ldÄ±!")
print("\nğŸ“‹ Ã–nemli Parametreler:")
print(f"   Epochs: {training_args.num_train_epochs}")
print(f"   Train Batch Size: {training_args.per_device_train_batch_size}")
print(f"   Eval Batch Size: {training_args.per_device_eval_batch_size}")
print(f"   Learning Rate: {training_args.learning_rate}")
print(f"   Warmup Steps: {training_args.warmup_steps}")
print(f"   Evaluation Strategy: {training_args.eval_strategy} (her {training_args.eval_steps} adÄ±m)")
print(f"   Save Strategy: {training_args.save_strategy} (her {training_args.save_steps} adÄ±m)")
print(f"   Best Model Metric: {training_args.metric_for_best_model}")

# Toplam adÄ±m tahmini
total_steps = (len(tokenized_dataset['train']) // training_args.per_device_train_batch_size) * training_args.num_train_epochs
print(f"\nğŸ“Š EÄŸitim Ä°statistikleri:")
print(f"   Toplam Training Ã–rnek: {len(tokenized_dataset['train']):,}")
print(f"   Toplam Validation Ã–rnek: {len(tokenized_dataset['validation']):,}")
print(f"   Batch per Epoch: {len(tokenized_dataset['train']) // training_args.per_device_train_batch_size}")
print(f"   Toplam Training Steps (tahmini): {total_steps:,}")
print(f"   Evaluation SayÄ±sÄ±: ~{total_steps // training_args.eval_steps}")

print("\n" + "="*60)
print("ğŸ¯ TÃœM MODEL VE LOGLARI BELÄ°RTÄ°LEN KLASÃ–RE KAYDEDÄ°LECEK!")
print("="*60)
````

````python
# ğŸ” Confusion Matrix Hesaplama ve GÃ¶rselleÅŸtirme

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from pathlib import Path

print("ğŸ” CONFUSION MATRIX ANALÄ°ZÄ°")
print("=" * 80)

# Confusion matrix hesapla
cm = confusion_matrix(test_labels, test_preds)

print("\nğŸ“Š Confusion Matrix:")
print(cm)

# Confusion matrix deÄŸerlerini aÃ§Ä±kla
tn, fp, fn, tp = cm.ravel()

print(f"\nğŸ“‹ Matris DeÄŸerleri:")
print(f"   True Negatives (TN):  {tn:,} - DoÄŸru Real tahminleri âœ…")
print(f"   False Positives (FP): {fp:,} - Real'i Fake diye tahmin âŒ")
print(f"   False Negatives (FN): {fn:,} - Fake'i Real diye tahmin âŒ")
print(f"   True Positives (TP):  {tp:,} - DoÄŸru Fake tahminleri âœ…")

# YÃ¼zdelik oranlar
total = tn + fp + fn + tp
print(f"\nğŸ“Š YÃ¼zdelik DaÄŸÄ±lÄ±m:")
print(f"   TN: {tn/total*100:.2f}%")
print(f"   FP: {fp/total*100:.2f}%")
print(f"   FN: {fn/total*100:.2f}%")
print(f"   TP: {tp/total*100:.2f}%")

# GÃ¶rselleÅŸtirme
fig, ax = plt.subplots(figsize=(10, 8))

# Confusion matrix display
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=['REAL', 'FAKE']
)

disp.plot(cmap='Blues', ax=ax, values_format='d', colorbar=True)
ax.set_title('Confusion Matrix - Fake News Detection\n', fontsize=16, fontweight='bold')
ax.set_xlabel('Predicted Label', fontsize=12, fontweight='bold')
ax.set_ylabel('True Label', fontsize=12, fontweight='bold')

# Her hÃ¼creye yÃ¼zde ekle
for i in range(2):
    for j in range(2):
        percentage = cm[i, j] / total * 100
        ax.text(j, i + 0.2, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=10, color='gray')

plt.tight_layout()

# GrafiÄŸi kaydet
plot_path = Path(OUTPUT_DIRS['plots']) / 'confusion_matrix.png'
plt.savefig(plot_path, dpi=300, bbox_inches='tight')
print(f"\nğŸ’¾ Grafik kaydedildi: {plot_path}")

plt.show()

# Hata analizi
print("\nâš ï¸  HATA ANALÄ°ZÄ°:")
if fp > fn:
    print(f"   Model daha Ã§ok False Positive yapÄ±yor ({fp} > {fn})")
    print("   â†’ Real haberleri Fake olarak iÅŸaretleme eÄŸilimi")
    print("   â†’ Daha tutucu (conservative) olabilir")
elif fn > fp:
    print(f"   Model daha Ã§ok False Negative yapÄ±yor ({fn} > {fp})")
    print("   â†’ Fake haberleri Real olarak kaÃ§Ä±rma eÄŸilimi")
    print("   â†’ Daha riskli! (Fake news yayÄ±labilir)")
else:
    print("   âœ… False Positive ve False Negative dengeli!")

print("=" * 80)
````

### ğŸ¨ 9.5 Prediction Confidence Analysis

Model tahminlerinin gÃ¼ven skorlarÄ±nÄ± inceleyelim.

````python
# ğŸ¨ Prediction Confidence Analysis

from scipy.special import softmax

print("ğŸ¨ PREDICTION CONFIDENCE ANALÄ°ZÄ°")
print("=" * 80)

# Logits'ten probabilities'e Ã§evir
probabilities = softmax(test_predictions.predictions, axis=1)

# Her tahmin iÃ§in max probability (confidence score)
confidence_scores = np.max(probabilities, axis=1)

print(f"\nğŸ“Š Confidence Score Ä°statistikleri:")
print(f"   Ortalama: {np.mean(confidence_scores):.4f}")
print(f"   Medyan:   {np.median(confidence_scores):.4f}")
print(f"   Min:      {np.min(confidence_scores):.4f}")
print(f"   Max:      {np.max(confidence_scores):.4f}")
print(f"   Std Dev:  {np.std(confidence_scores):.4f}")

# Confidence bins
bins = [0, 0.5, 0.7, 0.9, 0.95, 1.0]
bin_labels = ['<50%', '50-70%', '70-90%', '90-95%', '95-100%']
confidence_bins = pd.cut(confidence_scores, bins=bins, labels=bin_labels)

print(f"\nğŸ“Š Confidence DaÄŸÄ±lÄ±mÄ±:")
for label in bin_labels:
    count = (confidence_bins == label).sum()
    percentage = count / len(confidence_scores) * 100
    print(f"   {label}: {count:,} ({percentage:.2f}%)")

# GÃ¶rselleÅŸtirme
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# 1. Confidence histogram
axes[0].hist(confidence_scores, bins=30, color='#4ECDC4', edgecolor='black', alpha=0.7)
axes[0].axvline(np.mean(confidence_scores), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(confidence_scores):.3f}')
axes[0].set_title('Confidence Score Distribution', fontweight='bold', fontsize=14)
axes[0].set_xlabel('Confidence Score')
axes[0].set_ylabel('Frequency')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_facecolor('#F9F9F9')

# 2. DoÄŸru vs YanlÄ±ÅŸ tahminlerde confidence
correct_preds = test_preds == test_labels
correct_confidence = confidence_scores[correct_preds]
incorrect_confidence = confidence_scores[~correct_preds]

axes[1].hist(correct_confidence, bins=20, alpha=0.7, label=f'Correct ({len(correct_confidence):,})', color='#95E1D3')
axes[1].hist(incorrect_confidence, bins=20, alpha=0.7, label=f'Incorrect ({len(incorrect_confidence):,})', color='#FF6B6B')
axes[1].set_title('Confidence: Correct vs Incorrect Predictions', fontweight='bold', fontsize=14)
axes[1].set_xlabel('Confidence Score')
axes[1].set_ylabel('Frequency')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_facecolor('#F9F9F9')

plt.tight_layout()
plt.show()

# DÃ¼ÅŸÃ¼k confidence hatalarÄ± (en tehlikeli!)
low_conf_errors = (confidence_scores < 0.7) & (~correct_preds)
print(f"\nâš ï¸  DÃ¼ÅŸÃ¼k Confidence HatalarÄ±: {low_conf_errors.sum():,}")
print("   Bu Ã¶rnekler model iÃ§in belirsiz - incelenmeli!")

# YÃ¼ksek confidence hatalarÄ± (overconfident!)
high_conf_errors = (confidence_scores > 0.9) & (~correct_preds)
print(f"\nâš ï¸  YÃ¼ksek Confidence HatalarÄ±: {high_conf_errors.sum():,}")
print("   Model Ã§ok emin ama yanlÄ±ÅŸ - kritik hata!")

print("\nğŸ’¡ YBS Ä°Ã§gÃ¶rÃ¼sÃ¼:")
print("   Confidence score, modelin 'ne kadar emin' olduÄŸunu gÃ¶sterir.")
print("   DÃ¼ÅŸÃ¼k confidence â†’ 'Bilmiyorum' demek â†’ Ä°nsan review gerekli")
print("   YÃ¼ksek confidence + YanlÄ±ÅŸ â†’ Model bias'Ä± â†’ Tehlikeli!")

print("=" * 80)
````

### ğŸ“¦ 10.4 Inference Pipeline OluÅŸturma

Production'da kullanÄ±lmak Ã¼zere basit bir inference fonksiyonu oluÅŸturalÄ±m.

````python
# ğŸ“¦ Production-Ready Inference Pipeline

class FakeNewsDetector:
    """
    Fake News Detection iÃ§in production-ready inference sÄ±nÄ±fÄ±
    """

    def __init__(self, model_path, device='cpu'):
        """
        Args:
            model_path: Model dizini path
            device: 'cuda' veya 'cpu'
        """
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')

        # Model ve tokenizer yÃ¼kle
        self.model = AutoModelForSequenceClassification.from_pretrained(model_path)
        self.tokenizer = AutoTokenizer.from_pretrained(model_path)
        self.model = self.model.to(self.device)
        self.model.eval()

        # Label mapping
        self.id2label = {0: "REAL", 1: "FAKE"}

    def predict(self, text, return_probabilities=False):
        """
        Tek bir metin iÃ§in tahmin yap

        Args:
            text: Haber baÅŸlÄ±ÄŸÄ±
            return_probabilities: OlasÄ±lÄ±klarÄ± da dÃ¶ndÃ¼r

        Returns:
            dict: Tahmin sonuÃ§larÄ±
        """
        # Tokenize
        inputs = self.tokenizer(
            text,
            return_tensors='pt',
            padding=True,
            truncation=True,
            max_length=128
        )
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        # Predict
        with torch.no_grad():
            outputs = self.model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=-1)
            predicted_class = torch.argmax(probabilities, dim=-1).item()
            confidence = probabilities[0][predicted_class].item()

        result = {
            'text': text,
            'prediction': self.id2label[predicted_class],
            'confidence': confidence,
            'is_fake': predicted_class == 1
        }

        if return_probabilities:
            result['probabilities'] = {
                'REAL': probabilities[0][0].item(),
                'FAKE': probabilities[0][1].item()
            }

        return result

    def predict_batch(self, texts):
        """
        Birden fazla metin iÃ§in tahmin yap

        Args:
            texts: Liste of haber baÅŸlÄ±klarÄ±

        Returns:
            list: Tahmin sonuÃ§larÄ± listesi
        """
        return [self.predict(text) for text in texts]

# Detector oluÅŸtur
print("ğŸ“¦ INFERENCE PIPELINE OLUÅTURMA")
print("=" * 80)

detector = FakeNewsDetector(save_directory, device=str(device))
print("âœ… FakeNewsDetector oluÅŸturuldu!")

# Test Ã¶rnekleri
test_examples = [
    "Scientists discover cure for cancer",
    "Alien spaceship landed in New York City",
    "Stock market reaches new high",
    "President announces shocking conspiracy theory",
    "New research shows benefits of exercise"
]

print("\nğŸ§ª Test Ã–rnekleri:")
print("=" * 80)

for i, text in enumerate(test_examples, 1):
    result = detector.predict(text, return_probabilities=True)

    print(f"\n{i}. '{text}'")
    print(f"   Tahmin: {result['prediction']}")
    print(f"   GÃ¼ven: {result['confidence']:.4f} ({result['confidence']*100:.2f}%)")
    print(f"   OlasÄ±lÄ±klar:")
    print(f"      REAL: {result['probabilities']['REAL']:.4f}")
    print(f"      FAKE: {result['probabilities']['FAKE']:.4f}")

    # UyarÄ± sistemi
    if result['is_fake'] and result['confidence'] > 0.8:
        print(f"   âš ï¸  YÃœKSEK RÄ°SK: Fake olma ihtimali yÃ¼ksek!")
    elif result['confidence'] < 0.6:
        print(f"   âš¡ DÃœÅÃœK GÃœV EN: Ä°nsan review Ã¶nerilir")

print("\n" + "=" * 80)
print("âœ… Inference pipeline baÅŸarÄ±yla Ã§alÄ±ÅŸÄ±yor!")
print("=" * 80)
````
